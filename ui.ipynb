# Interactive DSA Trainer UI
Clean notebook without pre-saved widget state.


```
from core import ai_recommend_algorithm, ai_grade_code
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```
problem = widgets.Textarea(description='Problem')
code = widgets.Textarea(description='Code', layout=widgets.Layout(height='200px'))
output = widgets.Output()
```

```
def run_ai(b):
    with output:
        clear_output()
        print('Recommended:', ai_recommend_algorithm(problem.value))
        ok, msg = ai_grade_code(code.value, [([1,2,3], 6)])
        print(msg)

btn = widgets.Button(description='Run AI')
btn.on_click(run_ai)
display(problem, code, btn, output)
```
