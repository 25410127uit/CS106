```python
# @title üöÄ VISUALGO DSA
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import matplotlib.pyplot as plt
import networkx as nx
import random
import time
import re
import math


# ======================================================
# üÜô  D·ªÆ LI·ªÜU L√ù THUY·∫æT CHI TI·∫æT
# ======================================================
THEORY_CONTENT = {
    "Sorting": """
    <div style="font-family: 'Segoe UI', sans-serif; line-height: 1.6; font-size:14px;">
        <h3 style="color:#d9534f; border-bottom: 2px solid #d9534f;">CH∆Ø∆†NG 1: 6 THU·∫¨T TO√ÅN S·∫ÆP X·∫æP</h3>

        <details>
            <summary><b>1. Bubble Sort (N·ªïi b·ªçt) - O(n¬≤)</b></summary>
            <p><i>√ù t∆∞·ªüng:</i> Li√™n t·ª•c ƒë·ªïi ch·ªó 2 ph·∫ßn t·ª≠ k·ªÅ nhau n·∫øu ch√∫ng sai th·ª© t·ª±. Ph·∫ßn t·ª≠ l·ªõn nh·∫•t "n·ªïi" d·∫ßn v·ªÅ cu·ªëi.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #d9534f; padding:5px;">
for i = 0 to n-1:
    swapped = false
    for j = 0 to n-i-1:
        if A[j] > A[j+1]:
            swap(A[j], A[j+1])
            swapped = true
    if not swapped: break
            </pre>
        </details>

        <details>
            <summary><b>2. Selection Sort (Ch·ªçn) - O(n¬≤)</b></summary>
            <p><i>√ù t∆∞·ªüng:</i> T√¨m ph·∫ßn t·ª≠ nh·ªè nh·∫•t trong ƒëo·∫°n ch∆∞a s·∫Øp x·∫øp v√† ƒë∆∞a n√≥ v·ªÅ ƒë·∫ßu ƒëo·∫°n.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #d9534f; padding:5px;">
for i = 0 to n-1:
    min_idx = i
    for j = i+1 to n:
        if A[j] < A[min_idx]: min_idx = j
    swap(A[i], A[min_idx])
            </pre>
        </details>

        <details>
            <summary><b>3. Insertion Sort (Ch√®n) - O(n¬≤)</b></summary>
            <p><i>√ù t∆∞·ªüng:</i> Chia m·∫£ng l√†m 2 ph·∫ßn (ƒë√£ x·∫øp & ch∆∞a x·∫øp). L·∫•y t·ª´ng ph·∫ßn t·ª≠ ch√®n v√†o v·ªã tr√≠ ƒë√∫ng c·ªßa ph·∫ßn ƒë√£ x·∫øp.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #d9534f; padding:5px;">
for i = 1 to n:
    key = A[i]
    j = i - 1
    while j >= 0 and A[j] > key:
        A[j+1] = A[j]
        j--
    A[j+1] = key
            </pre>
        </details>

        <details>
            <summary><b>4. Merge Sort (Tr·ªôn) - O(n log n)</b></summary>
            <p><i>√ù t∆∞·ªüng (Chia ƒë·ªÉ tr·ªã):</i> Chia ƒë√¥i m·∫£ng ƒë·∫øn khi c√≤n 1 ph·∫ßn t·ª≠, sau ƒë√≥ tr·ªôn (Merge) c√°c m·∫£ng con l·∫°i theo th·ª© t·ª±.</p>
            <pre style="background:#e8f8f5; border-left:3px solid #1abc9c; padding:5px;">
func mergeSort(arr):
    if len(arr) <= 1: return arr
    mid = len(arr) // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    return merge(left, right)
            </pre>
        </details>

        <details>
            <summary><b>5. Quick Sort (Nhanh) - O(n log n)</b></summary>
            <p><i>√ù t∆∞·ªüng (Chia ƒë·ªÉ tr·ªã):</i> Ch·ªçn ch·ªët (Pivot). ƒê∆∞a c√°c s·ªë nh·ªè h∆°n Pivot sang tr√°i, l·ªõn h∆°n sang ph·∫£i.</p>
            <pre style="background:#e8f8f5; border-left:3px solid #1abc9c; padding:5px;">
func quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)
            </pre>
        </details>

        <details>
            <summary><b>6. Heap Sort (Vun ƒë·ªëng) - O(n log n)</b></summary>
            <p><i>√ù t∆∞·ªüng:</i> X√¢y d·ª±ng Max-Heap. Ph·∫ßn t·ª≠ l·ªõn nh·∫•t (g·ªëc) ƒë∆∞a v·ªÅ cu·ªëi, gi·∫£m k√≠ch th∆∞·ªõc Heap v√† vun l·∫°i.</p>
            <pre style="background:#e8f8f5; border-left:3px solid #1abc9c; padding:5px;">
func heapSort(arr):
    buildMaxHeap(arr)
    for i = n-1 down to 1:
        swap(arr[0], arr[i])
        heapify(arr, 0, i) // Vun l·∫°i ƒë·ªëng
            </pre>
        </details>
    </div>
    """,

    "LinkedList": """
    <div style="font-family: 'Segoe UI', sans-serif; line-height: 1.6;">
        <h3 style="color:#5cb85c; border-bottom: 2px solid #5cb85c;">CH∆Ø∆†NG 2: DANH S√ÅCH LI√äN K·∫æT (LINKED LIST)</h3>

        <details open>
            <summary><b>1. C·∫•u tr√∫c Node</b></summary>
            <pre style="background:#f4f4f4; border-left:3px solid #5cb85c; padding:5px;">
class Node {
    data: int      // D·ªØ li·ªáu
    next: Node* // Con tr·ªè tr·ªè ƒë·∫øn node k·∫ø ti·∫øp
}
            </pre>
        </details>

        <details>
            <summary><b>2. Th√™m sau Node Q (AddAfterQ) - O(n)</b></summary>
            <p>C·∫ßn duy·ªát t·ª´ ƒë·∫ßu ƒë·ªÉ t√¨m Q. N·∫øu th·∫•y Q th√¨ ch√®n node m·ªõi v√†o sau n√≥.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #5cb85c; padding:5px;">
func addAfterQ(head, q_val, new_val):
    curr = head
    while curr != null:
        if curr.data == q_val:
            newNode = new Node(new_val)
            newNode.next = curr.next
            curr.next = newNode
            return
        curr = curr.next
            </pre>
        </details>

        <details>
            <summary><b>3. Duy·ªát & T√¨m ki·∫øm (Traversal/Search)</b></summary>
            <p>Kh√°c v·ªõi m·∫£ng (truy c·∫≠p ng·∫´u nhi√™n O(1)), DSLK ph·∫£i truy c·∫≠p tu·∫ßn t·ª± O(n).</p>
        </details>
    </div>
    """,

    "BST": """
    <div style="font-family: 'Segoe UI', sans-serif; line-height: 1.6;">
        <h3 style="color:#f0ad4e; border-bottom: 2px solid #f0ad4e;">CH∆Ø∆†NG 3: C√ÇY NH·ªä PH√ÇN T√åM KI·∫æM (BST)</h3>

        <details>
            <summary><b>1. T√¨m ki·∫øm (Search) - O(log n)</b></summary>
            <p>T·∫≠n d·ª•ng t√≠nh ch·∫•t: Tr√°i < G·ªëc < Ph·∫£i ƒë·ªÉ lo·∫°i b·ªè m·ªôt n·ª≠a kh√¥ng gian t√¨m ki·∫øm m·ªói b∆∞·ªõc.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #f0ad4e; padding:5px;">
func search(root, key):
    if root is null or root.val == key: return root
    if key < root.val: return search(root.left, key)
    else: return search(root.right, key)
            </pre>
        </details>

        <details>
            <summary><b>2. Duy·ªát BFS (Chi·ªÅu r·ªông)</b></summary>
            <p>S·ª≠ d·ª•ng <b>Queue</b>. Duy·ªát l·∫ßn l∆∞·ª£t t·ª´ng t·∫ßng t·ª´ tr√™n xu·ªëng, tr√°i qua ph·∫£i.</p>
        </details>

        <details>
            <summary><b>3. Duy·ªát DFS (Chi·ªÅu s√¢u)</b></summary>
            <p>S·ª≠ d·ª•ng <b>Stack</b> (ho·∫∑c ƒê·ªá quy). G·ªìm 3 lo·∫°i:
            <ul>
                <li>Pre-order (Ti·ªÅn t·ª±): G·ªëc -> Tr√°i -> Ph·∫£i</li>
                <li>In-order (Trung t·ª±): Tr√°i -> G·ªëc -> Ph·∫£i (Ra d√£y tƒÉng d·∫ßn)</li>
                <li>Post-order (H·∫≠u t·ª±): Tr√°i -> Ph·∫£i -> G·ªëc</li>
            </ul>
            </p>
        </details>
    </div>
    """,

    "Graph": """
    <div style="font-family: 'Segoe UI', sans-serif; line-height: 1.6;">
        <h3 style="color:#5bc0de; border-bottom: 2px solid #5bc0de;">CH∆Ø∆†NG 4: L√ù THUY·∫æT ƒê·ªí TH·ªä & DIJKSTRA</h3>

        <details open>
            <summary><b>1. Thu·∫≠t to√°n Dijkstra (ƒê∆∞·ªùng ƒëi ng·∫Øn nh·∫•t)</b></summary>
            <p>T√¨m ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t t·ª´ ƒë·ªânh ngu·ªìn S ƒë·∫øn m·ªçi ƒë·ªânh kh√°c. Ch·ªâ √°p d·ª•ng cho tr·ªçng s·ªë d∆∞∆°ng.</p>
            <pre style="background:#f4f4f4; border-left:3px solid #5bc0de; padding:5px;">
func Dijkstra(Graph, S):
    D[v] = v√¥ c·ª±c v·ªõi m·ªçi v, D[S] = 0
    PQ = PriorityQueue()
    PQ.push(S, 0)

    while PQ not empty:
        u = PQ.pop_min()
        for v in neighbors(u):
            if D[u] + weight(u,v) < D[v]:
                D[v] = D[u] + weight(u,v)
                PQ.push(v, D[v])
            </pre>
        </details>

        <details>
            <summary><b>2. Bi·ªÉu di·ªÖn ƒê·ªì th·ªã</b></summary>
            <ul>
                <li><b>Ma tr·∫≠n k·ªÅ (Adjacency Matrix):</b> M·∫£ng 2 chi·ªÅu A[i][j]. T·ªën b·ªô nh·ªõ O(V¬≤). Nhanh khi ki·ªÉm tra c·∫°nh.</li>
                <li><b>Danh s√°ch k·ªÅ (Adjacency List):</b> M·∫£ng c√°c danh s√°ch. T·ªën b·ªô nh·ªõ O(V+E). T·ªët cho ƒë·ªì th·ªã th∆∞a.</li>
            </ul>
        </details>
    </div>
    """
}

# --- B. NG√ÇN H√ÄNG 10 C√ÇU H·ªéI TR·∫ÆC NGHI·ªÜM ---
QUIZ_DB = [
    {"q": "ƒê·ªô ph·ª©c t·∫°p th·ªùi gian t·ªìi nh·∫•t c·ªßa thu·∫≠t to√°n Bubble Sort l√† bao nhi√™u?",
     "opts": ["O(n)", "O(log n)", "O(n^2)", "O(1)"], "a": "O(n^2)"},

    {"q": "Trong danh s√°ch li√™n k·∫øt ƒë∆°n, thao t√°c AddHead (th√™m v√†o ƒë·∫ßu) t·ªën chi ph√≠ bao nhi√™u?",
     "opts": ["O(1)", "O(n)", "O(log n)", "O(n^2)"], "a": "O(1)"},

    {"q": "Thu·∫≠t to√°n t√¨m ki·∫øm theo chi·ªÅu r·ªông (BFS) s·ª≠ d·ª•ng c·∫•u tr√∫c d·ªØ li·ªáu n√†o?",
     "opts": ["Stack (NgƒÉn x·∫øp)", "Queue (H√†ng ƒë·ª£i)", "Array (M·∫£ng)", "Tree (C√¢y)"], "a": "Queue (H√†ng ƒë·ª£i)"},

    {"q": "Quy t·∫Øc s·∫Øp x·∫øp c·ªßa C√¢y nh·ªã ph√¢n t√¨m ki·∫øm (BST) l√† g√¨?",
     "opts": ["Tr√°i > G·ªëc > Ph·∫£i", "Ph·∫£i < G·ªëc < Tr√°i", "Tr√°i < G·ªëc < Ph·∫£i", "Kh√¥ng c√≥ quy t·∫Øc"], "a": "Tr√°i < G·ªëc < Ph·∫£i"},

    {"q": "Thu·∫≠t to√°n DFS (Depth-First Search) duy·ªát c√¢y theo th·ª© t·ª± n√†o?",
     "opts": ["Duy·ªát t·ª´ng t·∫ßng", "Duy·ªát s√¢u xu·ªëng nh√°nh con tr∆∞·ªõc", "Duy·ªát ng·∫´u nhi√™n", "Duy·ªát c√°c node l√° tr∆∞·ªõc"], "a": "Duy·ªát s√¢u xu·ªëng nh√°nh con tr∆∞·ªõc"},

    {"q": "C·∫•u tr√∫c d·ªØ li·ªáu n√†o ho·∫°t ƒë·ªông theo nguy√™n t·∫Øc LIFO (V√†o sau ra tr∆∞·ªõc)?",
     "opts": ["Queue", "Stack", "Linked List", "Graph"], "a": "Stack"},

    {"q": "ƒê·ªÉ x√≥a ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n c·ªßa Linked List, ta th·ª±c hi·ªán l·ªánh n√†o?",
     "opts": ["Head = Head.next", "Head = Null", "Head.next = Null", "Tail = Head"], "a": "Head = Head.next"},

    {"q": "N·∫øu c√¢y BST b·ªã l·ªách ho√†n to√†n v·ªÅ m·ªôt b√™n (nh∆∞ linked list), ƒë·ªô ph·ª©c t·∫°p t√¨m ki·∫øm l√†?",
     "opts": ["O(1)", "O(log n)", "O(n)", "O(n^2)"], "a": "O(n)"},

    {"q": "Thu·∫≠t to√°n t√¨m ki·∫øm nh·ªã ph√¢n (Binary Search) ch·ªâ √°p d·ª•ng ƒë∆∞·ª£c khi n√†o?",
     "opts": ["M·∫£ng ƒë√£ s·∫Øp x·∫øp", "M·∫£ng ng·∫´u nhi√™n", "Tr√™n Linked List", "M·∫£ng k√≠ch th∆∞·ªõc nh·ªè"], "a": "M·∫£ng ƒë√£ s·∫Øp x·∫øp"},

    {"q": "Trong AI, thu·∫≠t to√°n n√†o ƒë·∫£m b·∫£o t√¨m ƒë∆∞·ª£c ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t trong ƒë·ªì th·ªã kh√¥ng tr·ªçng s·ªë?",
     "opts": ["DFS", "BFS", "Greedy Best-First", "Random Walk"], "a": "BFS"}
]

# --- C. NG√ÇN H√ÄNG C√ÇU H·ªéI THI (POOL) ---
EXAM_QUESTION_POOL = [
    "Tr√¨nh b√†y √Ω t∆∞·ªüng v√† m√£ gi·∫£ c·ªßa thu·∫≠t to√°n Bubble Sort.",
    "So s√°nh ∆∞u nh∆∞·ª£c ƒëi·ªÉm c·ªßa Danh s√°ch li√™n k·∫øt (Linked List) so v·ªõi M·∫£ng (Array).",
    "V·∫Ω c√¢y Nh·ªã ph√¢n t√¨m ki·∫øm (BST) khi th√™m l·∫ßn l∆∞·ª£t c√°c gi√° tr·ªã: 50, 30, 20, 40, 70, 60, 80.",
    "Tr√¨nh b√†y s·ª± kh√°c bi·ªát gi·ªØa thu·∫≠t to√°n BFS v√† DFS. Khi n√†o n√™n d√πng BFS?",
    "Vi·∫øt h√†m AddAfterQ(Node Q, int value) trong danh s√°ch li√™n k·∫øt ƒë∆°n.",
    "Gi·∫£i th√≠ch kh√°i ni·ªám ƒê·ªô ph·ª©c t·∫°p thu·∫≠t to√°n (Big O). Cho v√≠ d·ª• v·ªÅ O(1) v√† O(n).",
    "T·∫°i sao c√¢y BST c√≥ th·ªÉ b·ªã suy bi·∫øn th√†nh danh s√°ch li√™n k·∫øt? L√†m sao ƒë·ªÉ kh·∫Øc ph·ª•c?",
    "M√¥ ph·ªèng ngƒÉn x·∫øp (Stack) b·∫±ng danh s√°ch li√™n k·∫øt: Vi·∫øt h√†m Push v√† Pop.",
    "Tr√¨nh b√†y thu·∫≠t to√°n T√¨m ki·∫øm nh·ªã ph√¢n (Binary Search). T·∫°i sao n√≥ nhanh h∆°n T√¨m ki·∫øm tuy·∫øn t√≠nh?",
    "Trong c√°c thu·∫≠t to√°n AI, Heuristic l√† g√¨? Cho v√≠ d·ª• trong b√†i to√°n t√¨m ƒë∆∞·ªùng.",
    "Vi·∫øt h√†m ƒë·∫øm s·ªë l∆∞·ª£ng node l√° trong c√¢y nh·ªã ph√¢n.",
    "Gi·∫£i th√≠ch c∆° ch·∫ø c·∫•p ph√°t b·ªô nh·ªõ ƒë·ªông c·ªßa Linked List."
]

# ==========================================
# 2. LOGIC & GIAO DI·ªÜN (SYSTEM)
# ==========================================
class TreeNode:
    def __init__(self, val): self.val=val; self.left=None; self.right=None

class VisualGo:
    def __init__(self):
        # UI Setup
        self.tabs = widgets.Tab()
        self.out_vis = widgets.Output()
        self.out_theory = widgets.Output()
        self.out_quiz = widgets.Output()
        self.out_assign = widgets.Output()
        self.out_exam = widgets.Output()
        self.out_ai = widgets.Output()
        self.tabs.children = [self.out_vis, self.out_theory, self.out_quiz, self.out_assign, self.out_exam, self.out_ai]
        for i, t in enumerate(['1. M√¥ ph·ªèng', '2. L√Ω thuy·∫øt', '3. Tr·∫Øc nghi·ªám', '4. B√†i t·∫≠p', '5. ƒê·ªÅ thi', '6. AI Tr·ª£ gi·∫£ng']):
            self.tabs.set_title(i, t)

        # Data Init
        self.sort_data = [15, 10, 5, 20, 25, 8]
        self.ll_data = [10, 20, 30, 40]
        self.bst_root = None
        self.init_bst()
        self.G = nx.Graph(); self.G.add_weighted_edges_from([('A','B',2),('B','C',1),('A','C',4),('C','D',3),('B','D',5)])

        display(HTML("<h2 style='text-align:center; color:#2c3e50;'>üéì Simulator Data Structures and Algorithms</h2>"))
        display(self.tabs)

        self.build_vis(); self.build_theory(); self.build_quiz(); self.build_assign(); self.build_exam(); self.build_ai()

    def init_bst(self):
        for x in [50, 30, 70, 20, 40, 60, 80]: self.bst_root = self.bst_insert(self.bst_root, x)

    # --- TAB 1: VISUALIZER ---
    def build_vis(self):
        with self.out_vis:
            menu = widgets.ToggleButtons(options=['Sorting', 'Linked List', 'BST', 'Graph'], button_style='info')
            scr = widgets.Output(layout={'border':'1px solid #ccc', 'height':'450px', 'padding':'10px'})
            ctl = widgets.Output(layout={'padding':'10px', 'background':'#f4f4f4'})
            # expose controls for cross-tab guidance (AI tr·ª£ gi·∫£ng)
            self.menu_main = menu
            self.scr_main = scr
            self.ctl_main = ctl

            def on_chg(c):
                with scr: clear_output();
                with ctl: clear_output();
                m = c['new']
                if m=='Sorting': self.ui_sort(scr, ctl)
                elif m=='Linked List': self.ui_ll(scr, ctl)
                elif m=='BST': self.ui_bst(scr, ctl)
                elif m=='Graph': self.ui_graph(scr, ctl)

            menu.observe(on_chg, names='value')
            display(menu, scr, ctl)
            self.ui_sort(scr, ctl)


    # MODULE 6: AI TR·ª¢ GI·∫¢NG (RAG-lite + Expert System)
    def build_ai(self):
        with self.out_ai:
            clear_output()
            display(HTML("<h3> AI Tr·ª£ gi·∫£ng (RAG-lite)</h3>"
                         "<p style='margin-top:0'>H·ªèi ƒë√°p theo l√Ω thuy·∫øt + t∆∞ v·∫•n ch·ªçn thu·∫≠t to√°n theo r√†ng bu·ªôc </p>"))
            # lazy init knowledge base
            self._ai_kb = None
            self.ui_ai()

    def _ai_strip_html(self, html: str) -> str:
        # b·ªè tag HTML c∆° b·∫£n ƒë·ªÉ l·∫•y text
        txt = re.sub(r'<[^>]+>', ' ', html)
        txt = re.sub(r'&nbsp;|&amp;|&lt;|&gt;', ' ', txt)
        txt = re.sub(r'\s+', ' ', txt).strip()
        return txt

    def _ai_tokenize(self, text: str):
        text = text.lower()
        text = re.sub(r'[^0-9a-z√†√°·∫°·∫£√£√¢·∫ß·∫•·∫≠·∫©·∫´ƒÉ·∫±·∫Ø·∫∑·∫≥·∫µ√®√©·∫π·∫ª·∫Ω√™·ªÅ·∫ø·ªá·ªÉ·ªÖ√¨√≠·ªã·ªâƒ©√≤√≥·ªç·ªè√µ√¥·ªì·ªë·ªô·ªï·ªó∆°·ªù·ªõ·ª£·ªü·ª°√π√∫·ª•·ªß≈©∆∞·ª´·ª©·ª±·ª≠·ªØ·ª≥√Ω·ªµ·ª∑·ªπƒë_ -]+', ' ', text)
        tokens = [t for t in text.split() if len(t) > 1]
        stop = set([
            # vi
            "l√†","v√†","c·ªßa","cho","trong","khi","m·ªôt","c√°c","nh·ªØng","ƒë∆∞·ª£c","t·ª´","ƒë·∫øn","v·ªõi","n√†y","ƒë√≥","th√¨","ta","s·∫Ω","n√™n","v√¨",
            "ho·∫∑c","nh∆∞","ƒë·ªÉ","theo","b·∫±ng","tr√™n","d∆∞·ªõi","c√≥","kh√¥ng","r·∫•t","nhi·ªÅu","√≠t","ƒëang",
            # en
            "the","a","an","and","or","of","to","in","on","for","with","is","are","be","as","by","from","this","that","it"
        ])
        return [t for t in tokens if t not in stop]

    def _ai_build_kb(self):
        # T·∫°o kho tri th·ª©c t·ª´ THEORY_CONTENT (chunk + TF-IDF)
        chunks = []
        for topic, html in THEORY_CONTENT.items():
            txt = self._ai_strip_html(html)
            # chia ƒëo·∫°n theo c√¢u/d·∫•u ch·∫•m ph·∫©y/ xu·ªëng d√≤ng l·ªõn
            parts = re.split(r'(?<=[\\.!\\?])\\s+|\\n\\s*\\n', txt)
            buf = []
            cur = ""
            for p in parts:
                p = p.strip()
                if not p:
                    continue
                if len(cur) < 400:
                    cur = (cur + " " + p).strip()
                else:
                    buf.append(cur); cur = p
            if cur:
                buf.append(cur)
            for c in buf:
                if len(c) >= 120:
                    chunks.append({"topic": topic, "text": c})

        # df
        df = {}
        docs_tokens = []
        for ch in chunks:
            toks = self._ai_tokenize(ch["text"])
            docs_tokens.append(toks)
            for t in set(toks):
                df[t] = df.get(t, 0) + 1

        N = max(1, len(chunks))
        idf = {t: math.log((N + 1) / (df_t + 1)) + 1.0 for t, df_t in df.items()}

        # vectorize chunks as dict + norm
        vecs = []
        for toks in docs_tokens:
            tf = {}
            for t in toks:
                tf[t] = tf.get(t, 0) + 1
            # tf-idf
            v = {t: (cnt / max(1, len(toks))) * idf.get(t, 0.0) for t, cnt in tf.items()}
            norm = math.sqrt(sum(w*w for w in v.values())) or 1.0
            vecs.append((v, norm))

        self._ai_kb = {"chunks": chunks, "idf": idf, "vecs": vecs}

    def _ai_cosine(self, v1, n1, v2, n2):
        # dot over intersection
        if len(v1) > len(v2):
            v1, v2 = v2, v1
            n1, n2 = n2, n1
        dot = 0.0
        for t, w in v1.items():
            if t in v2:
                dot += w * v2[t]
        return dot / (n1 * n2)

    def ai_retrieve(self, query: str, k: int = 3):
        if not self._ai_kb:
            self._ai_build_kb()
        toks = self._ai_tokenize(query)
        tf = {}
        for t in toks:
            tf[t] = tf.get(t, 0) + 1
        idf = self._ai_kb["idf"]
        qv = {t: (cnt / max(1, len(toks))) * idf.get(t, 0.0) for t, cnt in tf.items()}
        qn = math.sqrt(sum(w*w for w in qv.values())) or 1.0

        scored = []
        for i, (dv, dn) in enumerate(self._ai_kb["vecs"]):
            s = self._ai_cosine(qv, qn, dv, dn)
            scored.append((s, i))
        scored.sort(reverse=True, key=lambda x: x[0])
        results = []
        for s, i in scored[:k]:
            ch = self._ai_kb["chunks"][i]
            results.append({"score": s, "topic": ch["topic"], "text": ch["text"]})
        return results
    @staticmethod
    def clean_text(text):
        text = text.lower()
        text = re.sub(r"[^\w\s\-]", " ", text)
        text = re.sub(r"\s+", " ", text).strip()
        return text

    def ai_answer(self, question: str, mode: str = 'normal'):
        # AI tr·ª£ gi·∫£ng offline: router + template + n√∫t minh ho·∫°
        q_raw = question.strip()
        if not q_raw:
            return HTML("<i>Nh·∫≠p c√¢u h·ªèi tr∆∞·ªõc nh√©.</i>")
        q = q_raw.lower()

        def _mk_demo_button(topic_key: str):
            btn = widgets.Button(description="‚ñ∂Ô∏è Minh ho·∫° ngay", button_style='success', icon='play')
            out = widgets.Output()

            def _run(_):
                with out:
                    clear_output()
                    try:
                        # chuy·ªÉn sang tab m√¥ ph·ªèng
                        self.tabs.selected_index = 0
                        if not hasattr(self, 'menu_main'):
                            print("Kh√¥ng t√¨m th·∫•y v√πng m√¥ ph·ªèng.")
                            return

                        if topic_key in ('bubble', 'selection', 'insertion'):
                            self.menu_main.value = 'Sorting'
                            algo = {'bubble':'Bubble', 'selection':'Selection', 'insertion':'Insertion'}[topic_key]
                            self.run_sort(self.scr_main, algo)
                            print(f"ƒêang minh ho·∫° {algo} Sort‚Ä¶")
                        elif topic_key in ('bfs', 'dfs'):
                            self.menu_main.value = 'Graph'
                            start = sorted(list(self.G.nodes()))[0] if hasattr(self, 'G') else None
                            self.run_graph_traversal(self.scr_main, start, mode=topic_key)
                            print(f"ƒêang minh ho·∫° {topic_key.upper()}‚Ä¶")
                        else:
                            print("Ch∆∞a c√≥ m√¥ ph·ªèng tr·ª±c quan cho ch·ªß ƒë·ªÅ n√†y trong tab M√¥ ph·ªèng.")
                    except Exception as e:
                        print("L·ªói khi ch·∫°y minh ho·∫°:", e)

            btn.on_click(_run)
            return widgets.VBox([btn, out])

        def _style_answer(title, body_html, topic_key=None):
            head = f"<div style='font-size:16px'><b>ü§ñ {title}</b></div>"
            card = f"<div style='margin-top:8px; padding:10px; border-left:4px solid #3498db; background:#f7f9fb'>{body_html}</div>"
            if topic_key:
                return widgets.VBox([HTML(head + card), _mk_demo_button(topic_key)])
            return HTML(head + card)

        # ----- basic router -----
        topic_key = None
        if any(w in q for w in ["bubble", "n·ªïi b·ªçt", "noi bot"]):
            topic_key = "bubble"
        elif any(w in q for w in ["selection", "ch·ªçn"]):
            topic_key = "selection"
        elif any(w in q for w in ["insertion", "ch√®n", "chen"]):
            topic_key = "insertion"
        elif re.search(r"\bbfs\b|duy·ªát r·ªông|duyet rong", q):
            topic_key = "bfs"
        elif re.search(r"\bdfs\b|duy·ªát s√¢u|duyet sau", q):
            topic_key = "dfs"
        elif re.search(r"\bstack\b|ngƒÉn x·∫øp|ngan xep", q):
            topic_key = "stack"
        elif re.search(r"\bqueue\b|h√†ng ƒë·ª£i|hang doi", q):
            topic_key = "queue"

        intent = "general"
        if any(x in q for x in ["l√† g√¨", "la gi", "ƒë·ªãnh nghƒ©a", "define", "what is"]):
            intent = "define"
        elif any(x in q for x in ["kh√°c", "so s√°nh", "vs", "vs.", "ph√¢n bi·ªát", "phan biet"]):
            intent = "compare"
        elif any(x in q for x in ["khi n√†o", "nen dung", "n√™n d√πng", "use when", "tr∆∞·ªùng h·ª£p"]):
            intent = "when"
        elif any(x in q for x in ["ƒë·ªô ph·ª©c t·∫°p", "complexity", "big o", "o("]):
            intent = "complexity"

        # ----- templates -----
        def _ans_sort(name, idea, complexity, stable, in_place, when_use, code):
            if mode == "freshman":
                body = f"""<b>{name}</b><br>
    <b>Hi·ªÉu n√¥m na:</b> {idea}<br><br>
    <b>ƒê·ªô ph·ª©c t·∫°p:</b> {complexity}<br>
    <b>·ªîn ƒë·ªãnh (stable):</b> {stable} &nbsp; | &nbsp; <b>In-place:</b> {in_place}<br><br>
    <b>Khi n√™n d√πng:</b> {when_use}<br>
    <details style='margin-top:6px'><summary><b>M√£ gi·∫£</b></summary>
    <pre style='background:#fff; padding:8px; border:1px solid #eee'>{code}</pre></details>"""
            else:
                body = f"""<b>{name}</b>: {idea}<br>
    <ul>
      <li><b>Time:</b> {complexity}</li>
      <li><b>Stable:</b> {stable}</li>
      <li><b>In-place:</b> {in_place}</li>
      <li><b>Use when:</b> {when_use}</li>
    </ul>
    <details><summary><b>Pseudocode</b></summary>
    <pre style='background:#fff; padding:8px; border:1px solid #eee'>{code}</pre></details>"""
            return body

        T = {}

        T[("bubble","define")] = _ans_sort(
            "Bubble Sort (N·ªïi b·ªçt)",
            "ƒëi qua m·∫£ng nhi·ªÅu l·∫ßn, so s√°nh 2 ph·∫ßn t·ª≠ k·ªÅ nhau v√† ƒë·ªïi ch·ªó n·∫øu sai th·ª© t·ª±. Sau m·ªói v√≤ng, ph·∫ßn t·ª≠ l·ªõn nh·∫•t ‚Äún·ªïi‚Äù v·ªÅ cu·ªëi.",
            "O(n¬≤) worst/avg, O(n) best (n·∫øu c√≥ c·ªù swapped)",
            "C√≥ (n·∫øu ch·ªâ swap k·ªÅ nhau)",
            "C√≥",
            "m·∫£ng nh·ªè, ho·∫∑c d·ªØ li·ªáu g·∫ßn nh∆∞ ƒë√£ s·∫Øp x·∫øp v√† b·∫°n mu·ªën code c·ª±c ƒë∆°n gi·∫£n",
            "for i = 0..n-1:\n  swapped = false\n  for j = 0..n-2-i:\n    if a[j] > a[j+1]:\n      swap(a[j], a[j+1]); swapped = true\n  if not swapped: break"
        )

        T[("selection","define")] = _ans_sort(
            "Selection Sort (Ch·ªçn)",
            "m·ªói l∆∞·ª£t ch·ªçn ph·∫ßn t·ª≠ nh·ªè nh·∫•t trong ƒëo·∫°n ch∆∞a s·∫Øp x·∫øp r·ªìi ƒë∆∞a v·ªÅ ƒë·∫ßu ƒëo·∫°n ƒë√≥.",
            "O(n¬≤) m·ªçi tr∆∞·ªùng h·ª£p",
            "Kh√¥ng (th∆∞·ªùng kh√¥ng stable)",
            "C√≥",
            "khi mu·ªën √≠t swap, ho·∫∑c ƒë·ªÉ d·∫°y/hi·ªÉu √Ω t∆∞·ªüng ch·ªçn min",
            "for i = 0..n-1:\n  min = i\n  for j = i+1..n-1:\n    if a[j] < a[min]: min = j\n  swap(a[i], a[min])"
        )

        T[("insertion","define")] = _ans_sort(
            "Insertion Sort (Ch√®n)",
            "x√¢y d·∫ßn d√£y ƒë√£ s·∫Øp x·∫øp: l·∫•y ph·∫ßn t·ª≠ ti·∫øp theo v√† ch√®n v√†o ƒë√∫ng v·ªã tr√≠ trong ph·∫ßn b√™n tr√°i.",
            "O(n¬≤) worst, O(n) best (g·∫ßn sorted)",
            "C√≥",
            "C√≥",
            "d·ªØ li·ªáu g·∫ßn sorted; n nh·ªè; ho·∫∑c l√†m b∆∞·ªõc con trong TimSort",
            "for i = 1..n-1:\n  key = a[i]\n  j = i-1\n  while j>=0 and a[j] > key:\n    a[j+1] = a[j]; j--\n  a[j+1] = key"
        )

        if mode == "freshman":
            bfs_def = """<b>BFS</b> (duy·ªát r·ªông): ƒëi theo <b>t·ª´ng l·ªõp</b> nh∆∞ ‚Äús√≥ng lan‚Äù t·ª´ ƒëi·ªÉm b·∫Øt ƒë·∫ßu.
    D√πng <b>Queue</b>. Hay d√πng ƒë·ªÉ t√¨m <b>ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t tr√™n ƒë·ªì th·ªã kh√¥ng tr·ªçng s·ªë</b>."""
            dfs_def = """<b>DFS</b> (duy·ªát s√¢u): ƒëi <b>m·ªôt m·∫°ch th·∫≠t s√¢u</b> r·ªìi quay lui.
    D√πng <b>Stack</b> ho·∫∑c <b>ƒë·ªá quy</b>. Hay d√πng ƒë·ªÉ <b>ph√°t hi·ªán chu tr√¨nh</b>, <b>th√†nh ph·∫ßn li√™n th√¥ng</b>, topo sort (DAG)."""
        else:
            bfs_def = "<b>BFS</b>: level-order traversal using a <b>queue</b>; shortest path in unweighted graphs."
            dfs_def = "<b>DFS</b>: depth traversal using <b>stack/recursion</b>; cycles/CC/toposort use-cases."

        T[("bfs","define")] = f"""{bfs_def}
    <details><summary><b>Pseudocode</b></summary>
    <pre style='background:#fff; padding:8px; border:1px solid #eee'>queue = [start]; seen={{start}}\nwhile queue:\n  u = pop_left(queue)\n  for v in neighbors(u):\n    if v not in seen:\n      seen.add(v); push_right(queue, v)</pre></details>"""

        T[("dfs","define")] = f"""{dfs_def}
    <details><summary><b>Pseudocode</b></summary>
    <pre style='background:#fff; padding:8px; border:1px solid #eee'>stack=[start]; seen=set()\nwhile stack:\n  u=stack.pop()\n  if u in seen: continue\n  seen.add(u)\n  for v in neighbors(u):\n    if v not in seen: stack.append(v)</pre></details>"""

        compare_bfs_dfs = None
        if ("bfs" in q and "dfs" in q) or ("duy·ªát r·ªông" in q and "duy·ªát s√¢u" in q) or ("phan biet" in q and ("bfs" in q or "dfs" in q)):
            compare_bfs_dfs = """<b>BFS vs DFS</b><br>
    <ul>
      <li><b>√ù t∆∞·ªüng:</b> BFS ƒëi theo t·ª´ng l·ªõp; DFS ƒëi s√¢u r·ªìi quay lui.</li>
      <li><b>C·∫•u tr√∫c d·ªØ li·ªáu:</b> BFS d√πng <b>Queue</b>; DFS d√πng <b>Stack/Recursion</b>.</li>
      <li><b>ƒê∆∞·ªùng ƒëi ng·∫Øn (unweighted):</b> BFS ‚úÖ; DFS ‚ùå (kh√¥ng ƒë·∫£m b·∫£o).</li>
      <li><b>·ª®ng d·ª•ng:</b> DFS m·∫°nh cho cycle/connected components/topo sort; BFS m·∫°nh cho shortest path kh√¥ng tr·ªçng s·ªë.</li>
    </ul>"""

        if mode == "freshman":
            stack_def = """<b>Stack (NgƒÉn x·∫øp)</b>: gi·ªëng ch·ªìng ƒëƒ©a. V√†o sau ra tr∆∞·ªõc (<b>LIFO</b>).<br>
    Thao t√°c: <b>push</b> (th√™m), <b>pop</b> (l·∫•y ra), <b>top</b> (xem ƒë·ªânh)."""
            queue_def = """<b>Queue (H√†ng ƒë·ª£i)</b>: gi·ªëng x·∫øp h√†ng mua v√©. V√†o tr∆∞·ªõc ra tr∆∞·ªõc (<b>FIFO</b>).<br>
    Thao t√°c: <b>enqueue</b> (th√™m v√†o cu·ªëi), <b>dequeue</b> (l·∫•y ·ªü ƒë·∫ßu), <b>front</b>."""
        else:
            stack_def = "<b>Stack</b>: LIFO; push/pop/top."
            queue_def = "<b>Queue</b>: FIFO; enqueue/dequeue/front."

        T[("stack","define")] = stack_def + "<br><b>V√≠ d·ª•:</b> DFS, undo/redo, call stack."
        T[("queue","define")] = queue_def + "<br><b>V√≠ d·ª•:</b> BFS, x·ª≠ l√Ω t√°c v·ª• theo l∆∞·ª£t (scheduler)."

        # fallback to RAG-lite
        def _rag():
            hits = self.ai_retrieve(q_raw, k=3)
            if not hits or hits[0]["score"] < 0.05:
                return HTML("<b> G·ª£i √Ω:</b> M√¨nh ch∆∞a t√¨m th·∫•y ƒëo·∫°n l√Ω thuy·∫øt kh·ªõp. B·∫°n th·ª≠ th√™m t·ª´ kh√≥a (vd: 'BFS', 'heap', 'linked list', 'ƒë·ªô ph·ª©c t·∫°p').")
            html = "<b> Tr·∫£ l·ªùi g·ª£i √Ω (truy xu·∫•t t·ª´ l√Ω thuy·∫øt):</b><br>"
            for h in hits:
                excerpt = (h["text"][:420] + "‚Ä¶") if len(h["text"]) > 420 else h["text"]
                html += f"<div style='margin:10px 0; padding:10px; border-left:4px solid #2c3e50; background:#f7f9fb'><b>Ch·ªß ƒë·ªÅ:</b> {h['topic']} &nbsp; <span style='color:#888'>(score={h['score']:.2f})</span><br>{excerpt}</div>"
            html += "<div style='color:#888; font-size:12px'>*Ghi ch√∫: ƒê√¢y l√† RAG-lite ch·∫°y offline (truy xu·∫•t ƒëo·∫°n li√™n quan), kh√¥ng ph·∫£i LLM online.</div>"
            return HTML(html)

        if intent == "compare" and compare_bfs_dfs:
            return _style_answer("So s√°nh", compare_bfs_dfs, topic_key="bfs")

        key = (topic_key, "define")
        if topic_key and key in T:
            title = "Gi·∫£i th√≠ch nh∆∞ SV nƒÉm 1" if mode == "freshman" else "Tr·∫£ l·ªùi"
            demo_key = topic_key if topic_key in ('bubble','selection','insertion','bfs','dfs') else None
            return _style_answer(title, T[key], topic_key=demo_key)

        return _rag()



    def ai_recommend(self, task: str, **kwargs):
        # Expert-system rules
        if task == "Sorting":
            n = kwargs.get("n", 100)
            nearly_sorted = kwargs.get("nearly_sorted", False)
            stable = kwargs.get("stable", False)
            memory_limited = kwargs.get("memory_limited", False)
            range_small = kwargs.get("range_small", False)

            if range_small:
                return ("Counting Sort / Radix (n·∫øu s·ªë nguy√™n)",
                        "Mi·ªÅn gi√° tr·ªã nh·ªè -> ƒë·∫øm (Counting) ho·∫∑c Radix cho s·ªë nguy√™n; th·ªùi gian g·∫ßn O(n).")
            if nearly_sorted:
                return ("Insertion Sort",
                        "G·∫ßn nh∆∞ ƒë√£ s·∫Øp x·∫øp -> Insertion Sort ch·∫°y r·∫•t nhanh (g·∫ßn O(n)).")
            if stable and not memory_limited:
                return ("Merge Sort",
                        "C·∫ßn ·ªïn ƒë·ªãnh (stable) -> Merge Sort ·ªïn ƒë·ªãnh, O(n log n) nh∆∞ng t·ªën b·ªô nh·ªõ ph·ª•.")
            if memory_limited:
                return ("Heap Sort",
                        "Gi·ªõi h·∫°n b·ªô nh·ªõ -> Heap Sort in-place, O(n log n) (kh√¥ng stable).")
            if n > 5000:
                return ("Quick Sort (random/median pivot)",
                        "D·ªØ li·ªáu l·ªõn -> Quick Sort th∆∞·ªùng nhanh trong th·ª±c t·∫ø; ch·ªçn pivot t·ªët ƒë·ªÉ tr√°nh worst-case.")
            return ("Quick Sort / Merge Sort", "T·ªïng qu√°t: Quick (nhanh) ho·∫∑c Merge (stable).")

        if task == "Graph":
            weighted = kwargs.get("weighted", False)
            negative = kwargs.get("negative", False)
            need_mst = kwargs.get("need_mst", False)
            shortest = kwargs.get("shortest", True)

            if need_mst:
                return ("Kruskal / Prim", "C·∫ßn c√¢y khung nh·ªè nh·∫•t (MST) -> d√πng Kruskal ho·∫∑c Prim.")
            if shortest:
                if not weighted:
                    return ("BFS", "ƒê·ªì th·ªã kh√¥ng tr·ªçng s·ªë -> BFS ƒë·∫£m b·∫£o ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t theo s·ªë c·∫°nh.")
                if weighted and not negative:
                    return ("Dijkstra", "Tr·ªçng s·ªë kh√¥ng √¢m -> Dijkstra t·ªëi ∆∞u ƒë∆∞·ªùng ƒëi ng·∫Øn.")
                if negative:
                    return ("Bellman-Ford", "C√≥ c·∫°nh √¢m -> Bellman-Ford (ph√°t hi·ªán chu tr√¨nh √¢m).")
            return ("DFS", "C·∫ßn duy·ªát/ki·ªÉm tra li√™n th√¥ng/chu tr√¨nh -> DFS ph√π h·ª£p.")

        if task == "Tree":
            ops = kwargs.get("ops", "search")
            need_balance = kwargs.get("need_balance", True)
            if ops in ("priority", "k_smallest"):
                return ("Heap (Priority Queue)", "∆Øu ti√™n l·∫•y min/max li√™n t·ª•c -> Heap ph√π h·ª£p.")
            if need_balance:
                return ("AVL / Red-Black Tree", "C·∫ßn t√¨m ki·∫øm/insert/delete nhanh ·ªïn ƒë·ªãnh -> c√¢y t·ª± c√¢n b·∫±ng (AVL/RB).")
            return ("BST", "ƒê∆°n gi·∫£n, d·ªÖ c√†i; nh∆∞ng c√≥ th·ªÉ l·ªách -> O(n) worst-case.")

        if task == "Linked List":
            frequent_head = kwargs.get("frequent_head", True)
            need_bidirectional = kwargs.get("need_bidirectional", False)
            if need_bidirectional:
                return ("Doubly Linked List", "C·∫ßn duy·ªát ng∆∞·ª£c / x√≥a node khi c√≥ con tr·ªè -> d√πng doubly linked list.")
            if frequent_head:
                return ("Singly Linked List", "Ch√®n/x√≥a ƒë·∫ßu danh s√°ch O(1) -> singly linked list ph√π h·ª£p.")
            return ("Linked List + Tail pointer", "Hay th√™m cu·ªëi -> gi·ªØ con tr·ªè tail ƒë·ªÉ push O(1).")

        return ("N/A", "Ch∆∞a h·ªó tr·ª£ lo·∫°i b√†i to√°n n√†y.")

    def ai_grade_code(self, task_key: str, user_code: str):
        """Ch·∫•m b√†i offline: ch·∫°y test + ph√¢n t√≠ch m√£ (rule-based) -> tr·∫£ HTML report."""
        import ast, traceback
        def html_escape(s: str):
            return (s.replace("&","&amp;").replace("<","&lt;").replace(">","&gt;"))

        issues = []
        tips = []
        score = 100

        # --- Quick static checks (kh√¥ng c·∫ßn ch·∫°y code) ---
        lowered = user_code.lower()
        if "import os" in lowered or "import sys" in lowered or "subprocess" in lowered:
            issues.append("‚õî Kh√¥ng cho ph√©p d√πng os/sys/subprocess trong b√†i ch·∫•m t·ª± ƒë·ªông.")
            score -= 40
        if "open(" in lowered or "eval(" in lowered or "__import__" in lowered:
            issues.append("‚õî Kh√¥ng cho ph√©p d√πng open/eval/__import__ trong m√¥i tr∆∞·ªùng ch·∫•m.")
            score -= 40

        # Bubble Sort: ch·∫∑n d√πng built-in sort
        if task_key == "bubble_sort":
            if "sorted(" in lowered or ".sort(" in lowered:
                issues.append("‚ö†Ô∏è B·∫°n ƒëang d√πng sorted()/list.sort(). B√†i y√™u c·∫ßu t·ª± c√†i ƒë·∫∑t Bubble Sort.")
                score -= 40
                tips.append("G·ª£i √Ω: d√πng 2 v√≤ng l·∫∑p, so s√°nh c·∫∑p k·ªÅ nhau v√† swap n·∫øu sai th·ª© t·ª±.")

        # Parse AST ƒë·ªÉ b·∫Øt l·ªói c√∫ ph√°p s·ªõm
        try:
            ast.parse(user_code)
        except SyntaxError as e:
            score = 0
            return HTML(
                f"<div style='font-family:Segoe UI'>"
                f"<h4 style='margin:0;color:#d9534f'>‚ùå L·ªói c√∫ ph√°p</h4>"
                f"<pre style='background:#fff3f3;padding:10px;border:1px solid #f5c2c7'>{html_escape(str(e))}</pre>"
                f"</div>"
            )

        # --- Execute in a restricted-ish scope ---
        safe_builtins = {
            "abs": abs, "all": all, "any": any, "enumerate": enumerate, "len": len,
            "list": list, "dict": dict, "set": set, "tuple": tuple, "range": range,
            "min": min, "max": max, "sum": sum, "print": print,
        }
        g = {"__builtins__": safe_builtins}
        l = {}
        try:
            exec(user_code, g, l)
        except Exception as e:
            score = max(0, score-50)
            tb = traceback.format_exc(limit=2)
            issues.append("üí• Code ch·∫°y b·ªã l·ªói runtime khi import/kh·ªüi t·∫°o.")
            return HTML(
                f"<div style='font-family:Segoe UI'>"
                f"<h4 style='margin:0;color:#d9534f'>‚ùå Runtime error</h4>"
                f"<p>ƒêi·ªÉm t·∫°m t√≠nh: <b>{max(0,score)}</b>/100</p>"
                f"<pre style='background:#f7f7f7;padding:10px;border:1px solid #ddd'>{html_escape(tb)}</pre>"
                f"</div>"
            )

        # --- Tests ---
        import random
        passed = 0
        total = 0

        def add_issue(msg, penalty=10):
            nonlocal score
            issues.append(msg)
            score -= penalty

        if task_key == "bubble_sort":
            fn = l.get("bubble_sort") or g.get("bubble_sort")
            if not callable(fn):
                return HTML("<b style='color:#d9534f'>‚ùå Kh√¥ng t√¨m th·∫•y h√†m bubble_sort(arr).</b>")
            for _ in range(12):
                total += 1
                arr = [random.randint(-30, 30) for _ in range(random.randint(0, 25))]
                exp = sorted(arr)
                try:
                    out = fn(list(arr))
                    # ch·∫•p nh·∫≠n tr·∫£ v·ªÅ arr ho·∫∑c s·ª≠a in-place
                    if out is None:
                        # assume in-place; but we passed a copy so can't read
                        add_issue("‚ö†Ô∏è H√†m tr·∫£ v·ªÅ None. N√™n return arr (ƒë·ªÉ d·ªÖ test).", penalty=5)
                        continue
                    if out == exp:
                        passed += 1
                    else:
                        add_issue(f"‚ùå Sai test: input={arr} -> output={out}, expected={exp}", penalty=6)
                except Exception:
                    add_issue("üí• L·ªói runtime khi ch·∫°y bubble_sort tr√™n test ng·∫´u nhi√™n.", penalty=12)
            if passed == total:
                tips.append("‚úÖ ƒê√∫ng t·∫•t c·∫£ test. B·∫°n c√≥ th·ªÉ t·ªëi ∆∞u b·∫±ng c·ªù 'swapped' ƒë·ªÉ d·ª´ng s·ªõm khi m·∫£ng ƒë√£ sorted.")

        elif task_key == "ll_add_head":
            # expect Node and LinkedList.add_head
            LinkedList = l.get("LinkedList") or g.get("LinkedList")
            Node = l.get("Node") or g.get("Node")
            if LinkedList is None or Node is None:
                return HTML("<b style='color:#d9534f'>‚ùå C·∫ßn c√≥ class Node v√† class LinkedList.</b>")
            ll_obj = LinkedList()
            if not hasattr(ll_obj, "add_head"):
                return HTML("<b style='color:#d9534f'>‚ùå Kh√¥ng t√¨m th·∫•y method LinkedList.add_head(self, data).</b>")
            try:
                ll_obj.add_head(10)
                total += 1
                if ll_obj.head is None or getattr(ll_obj.head, "data", None) != 10:
                    add_issue("‚ùå Sau add_head(10), head ph·∫£i ch·ª©a data=10.", penalty=15)
                else:
                    passed += 1
                ll_obj.add_head(20)
                total += 1
                if ll_obj.head is None or getattr(ll_obj.head, "data", None) != 20:
                    add_issue("‚ùå Sau add_head(20), head ph·∫£i l√† 20.", penalty=15)
                else:
                    # head.next should be previous 10
                    nxt = getattr(ll_obj.head, "next", None)
                    if nxt is None or getattr(nxt, "data", None) != 10:
                        add_issue("‚ùå Con tr·ªè next c·ªßa head ph·∫£i tr·ªè t·ªõi node c≈© (10).", penalty=12)
                    else:
                        passed += 1
                        tips.append("‚úÖ Add head chu·∫©n: new_node.next = head; head = new_node.")
            except Exception:
                add_issue("üí• L·ªói runtime khi g·ªçi add_head.", penalty=20)

        elif task_key == "bfs":
            bfs = l.get("bfs") or g.get("bfs")
            if not callable(bfs):
                return HTML("<b style='color:#d9534f'>‚ùå Kh√¥ng t√¨m th·∫•y h√†m bfs(adj, start).</b>")
            # Deterministic test graphs
            tests = [
                ({0:[1,2], 1:[3], 2:[3], 3:[]}, 0, [0,1,2,3]),
                ({'A':['B','C'], 'B':['D'], 'C':['D'], 'D':[]}, 'A', ['A','B','C','D'])
            ]
            for adj, start, exp in tests:
                total += 1
                try:
                    out = bfs(adj, start)
                    if out == exp:
                        passed += 1
                    else:
                        add_issue(f"‚ùå Sai BFS: start={start}, output={out}, expected={exp}", penalty=15)
                except Exception:
                    add_issue("üí• L·ªói runtime khi ch·∫°y bfs.", penalty=20)
            if "deque" not in lowered:
                tips.append("G·ª£i √Ω: BFS th∆∞·ªùng d√πng h√†ng ƒë·ª£i (queue). Python: collections.deque.")
        else:
            return HTML("<b style='color:#d9534f'>‚ùå B√†i ch·∫•m ch∆∞a h·ªó tr·ª£.</b>")

        score = max(0, min(100, score))
        pct = (passed/total*100) if total else 0
        color = "#28a745" if score >= 80 else ("#f0ad4e" if score >= 50 else "#d9534f")

        def bullet(items):
            if not items:
                return "<i>Kh√¥ng c√≥.</i>"
            return "<ul>" + "".join([f"<li>{html_escape(x)}</li>" for x in items]) + "</ul>"

        return HTML(
            f"<div style='font-family:Segoe UI; line-height:1.55'>"
            f"<h4 style='margin:0'>üßæ B√°o c√°o ch·∫•m</h4>"
            f"<p style='margin:6px 0'>Test pass: <b>{passed}/{total}</b> ({pct:.0f}%) &nbsp; | &nbsp; ƒêi·ªÉm: "
            f"<b style='color:{color}'>{score}/100</b></p>"
            f"<div style='display:grid; grid-template-columns:1fr 1fr; gap:12px'>"
            f"<div style='padding:10px; border:1px solid #ddd; border-radius:10px'>"
            f"<b>V·∫•n ƒë·ªÅ ph√°t hi·ªán</b>{bullet(issues)}"
            f"</div>"
            f"<div style='padding:10px; border:1px solid #ddd; border-radius:10px'>"
            f"<b>G·ª£i √Ω c·∫£i thi·ªán</b>{bullet(tips)}"
            f"</div>"
            f"</div>"
            f"<p style='color:#777; font-size:12px; margin-top:8px'>*Ch·∫•m offline: test + lu·∫≠t kinh nghi·ªám (kh√¥ng g·ªçi Internet).</p>"
            f"</div>"
        )


    def ui_ai(self):
        # --- Tab 1: H·ªèi ƒë√°p
        q_input = widgets.Textarea(
            placeholder="V√≠ d·ª•: 'BFS kh√°c DFS ·ªü ƒëi·ªÉm n√†o?' ho·∫∑c 'T·∫°i sao Insertion Sort nhanh khi g·∫ßn sorted?'",
            layout=widgets.Layout(width='100%', height='90px')
        )
        mode_toggle = widgets.ToggleButtons(
            options=[('Chu·∫©n', 'normal'), ('SV nƒÉm 1', 'freshman')],
            value='freshman',
            button_style='',
            tooltips=['Tr·∫£ l·ªùi ng·∫Øn g·ªçn, ƒë√∫ng tr·ªçng t√¢m', 'Gi·∫£i th√≠ch d·ªÖ hi·ªÉu nh∆∞ cho sinh vi√™n nƒÉm 1']
        )
        btn_ask = widgets.Button(description="üí¨ H·ªèi AI", button_style='info')
        out_ans = widgets.Output()

        def on_ask(_):
            with out_ans:
                clear_output()
                if not q_input.value.strip():
                    display(HTML("<i>Nh·∫≠p c√¢u h·ªèi tr∆∞·ªõc nh√©.</i>"))
                    return
                self.ai_mode = mode_toggle.value
                display(self.ai_answer(q_input.value.strip(), mode=self.ai_mode))

        btn_ask.on_click(on_ask)
        tab1 = widgets.VBox([q_input, mode_toggle, btn_ask, out_ans])

        # --- Tab 2: T∆∞ v·∫•n ch·ªçn thu·∫≠t to√°n (Expert System)
        task_dd = widgets.Dropdown(options=["Sorting", "Graph", "Tree", "Linked List"], value="Sorting", description="B√†i to√°n:")
        form_box = widgets.VBox()
        btn_rec = widgets.Button(description="üß† T∆∞ v·∫•n", button_style='success')
        out_rec = widgets.Output()

        # Sorting widgets
        w_n = widgets.IntSlider(value=200, min=10, max=20000, step=10, description="n:", continuous_update=False)
        w_near = widgets.Checkbox(value=False, description="G·∫ßn nh∆∞ sorted")
        w_stable = widgets.Checkbox(value=False, description="C·∫ßn stable")
        w_mem = widgets.Checkbox(value=False, description="Gi·ªõi h·∫°n b·ªô nh·ªõ")
        w_range = widgets.Checkbox(value=False, description="Mi·ªÅn gi√° tr·ªã nh·ªè")

        # Graph widgets
        w_weighted = widgets.Checkbox(value=False, description="C√≥ tr·ªçng s·ªë")
        w_negative = widgets.Checkbox(value=False, description="C√≥ c·∫°nh √¢m")
        w_mst = widgets.Checkbox(value=False, description="C·∫ßn MST")
        w_shortest = widgets.Checkbox(value=True, description="C·∫ßn ƒë∆∞·ªùng ƒëi ng·∫Øn")

        # Tree widgets
        w_ops = widgets.Dropdown(options=["search", "insert", "delete", "traversal", "k_smallest"], value="search", description="M·ª•c ti√™u:")
        w_balance = widgets.Checkbox(value=True, description="C·∫ßn c√¢n b·∫±ng")

        # Linked List widgets
        w_head = widgets.Checkbox(value=True, description="Hay ch√®n/x√≥a ƒë·∫ßu")
        w_bi = widgets.Checkbox(value=False, description="C·∫ßn duy·ªát 2 chi·ªÅu")

        def render_form(task):
            if task == "Sorting":
                form_box.children = [w_n, w_near, w_stable, w_mem, w_range]
            elif task == "Graph":
                form_box.children = [w_weighted, w_negative, w_mst, w_shortest]
            elif task == "Tree":
                form_box.children = [w_ops, w_balance]
            else:
                form_box.children = [w_head, w_bi]

        render_form(task_dd.value)

        def on_task_change(change):
            render_form(change["new"])
            with out_rec:
                clear_output()
        task_dd.observe(on_task_change, names='value')

        def on_rec(_):
            with out_rec:
                clear_output()
                task = task_dd.value
                if task == "Sorting":
                    algo, reason = self.ai_recommend("Sorting", n=w_n.value, nearly_sorted=w_near.value, stable=w_stable.value,
                                                    memory_limited=w_mem.value, range_small=w_range.value)
                elif task == "Graph":
                    algo, reason = self.ai_recommend("Graph", weighted=w_weighted.value, negative=w_negative.value,
                                                    need_mst=w_mst.value, shortest=w_shortest.value)
                elif task == "Tree":
                    algo, reason = self.ai_recommend("Tree", ops=w_ops.value, need_balance=w_balance.value)
                else:
                    algo, reason = self.ai_recommend("Linked List", frequent_head=w_head.value, need_bidirectional=w_bi.value)

                display(HTML(f"<h4>‚úÖ G·ª£i √Ω thu·∫≠t to√°n: <span style='color:#2c3e50'>{algo}</span></h4>"
                             f"<div style='padding:10px; background:#f7f9fb; border-left:4px solid #2c3e50'>{reason}</div>"))

        btn_rec.on_click(on_rec)
        tab2 = widgets.VBox([task_dd, form_box, btn_rec, out_rec])

        # --- Tab 3: Quiz nhanh (auto)
        topic_dd = widgets.Dropdown(options=list(THEORY_CONTENT.keys()), description="Ch·ªß ƒë·ªÅ:")
        btn_q = widgets.Button(description="üéØ Sinh c√¢u h·ªèi", button_style='warning')
        out_q = widgets.Output()

        def gen_quiz(_):
            with out_q:
                clear_output()
                topic = topic_dd.value
                # t·∫°o c√¢u h·ªèi ƒë∆°n gi·∫£n d·ª±a tr√™n m·∫´u
                templates = [
                    ("ƒê·ªô ph·ª©c t·∫°p trung b√¨nh c·ªßa Quick Sort l√† g√¨?", ["O(n)", "O(n log n)", "O(log n)", "O(n^2)"], "O(n log n)"),
                    ("BFS ƒë·∫£m b·∫£o ƒëi·ªÅu g√¨ trong ƒë·ªì th·ªã kh√¥ng tr·ªçng s·ªë?", ["T√¨m ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t", "Lu√¥n duy·ªát s√¢u nh·∫•t", "T·ªëi ∆∞u MST", "Ph√°t hi·ªán c·∫°nh √¢m"], "T√¨m ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t"),
                    ("Trong Linked List, thao t√°c n√†o th∆∞·ªùng O(1)?", ["Truy c·∫≠p ph·∫ßn t·ª≠ k", "Ch√®n ƒë·∫ßu danh s√°ch", "T√¨m ki·∫øm gi√° tr·ªã", "S·∫Øp x·∫øp"], "Ch√®n ƒë·∫ßu danh s√°ch"),
                    ("Heap th∆∞·ªùng d√πng ƒë·ªÉ l√†m g√¨?", ["Priority Queue", "L∆∞u ƒë·ªì th·ªã", "Hashing", "N√©n d·ªØ li·ªáu"], "Priority Queue"),
                ]
                q, opts, ans = random.choice(templates)
                display(HTML(f"<b>C√¢u h·ªèi ({topic}):</b> {q}"))
                radios = widgets.RadioButtons(options=opts)
                btn_check = widgets.Button(description="‚úÖ Ki·ªÉm tra", button_style='success')
                out_fb = widgets.Output()

                def check(_):
                    with out_fb:
                        clear_output()
                        if radios.value == ans:
                            display(HTML("<b style='color:green'>ƒê√öNG ‚úÖ</b>"))
                        else:
                            display(HTML(f"<b style='color:red'>SAI ‚ùå</b> ƒê√°p √°n ƒë√∫ng: <b>{ans}</b>"))

                btn_check.on_click(check)
                display(radios, btn_check, out_fb)

        btn_q.on_click(gen_quiz)
        # --- Tab 4: Ch·∫•m b√†i + ph·∫£n h·ªìi (AI offline)
        grade_task = widgets.Dropdown(
            options=[
                ("Bubble Sort (def bubble_sort(arr))", "bubble_sort"),
                ("Linked List (def add_head(self, data))", "ll_add_head"),
                ("BFS tr√™n c√¢y/ƒë·ªì th·ªã (def bfs(...))", "bfs"),
            ],
            value="bubble_sort",
            description="B√†i:"
        )
        code_tpl = {
            "bubble_sort": '''def bubble_sort(arr):
    # TODO: s·∫Øp x·∫øp tƒÉng d·∫ßn (in-place)
    return arr
''',
            "ll_add_head": '''class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def add_head(self, data):
        # TODO: th√™m node v√†o ƒë·∫ßu danh s√°ch
        pass
''',
            "bfs": '''from collections import deque

def bfs(adj, start):
    # adj: dict {u:[v1,v2,...]}
    # TODO: tr·∫£ v·ªÅ danh s√°ch th·ª© t·ª± duy·ªát BFS
    return []
''',
        }
        code_in = widgets.Textarea(value=code_tpl["bubble_sort"], layout=widgets.Layout(width="100%", height="220px"))
        btn_grade = widgets.Button(description="üß™ Ch·∫•m & g√≥p √Ω", button_style="warning")
        out_grade = widgets.Output()

        def on_task(_):
            code_in.value = code_tpl[grade_task.value]
            with out_grade:
                clear_output()
                display(HTML("<i>D√°n/vi·∫øt l·ªùi gi·∫£i r·ªìi b·∫•m <b>Ch·∫•m & g√≥p √Ω</b>. H·ªá th·ªëng ch·∫•m offline (test + ph√¢n t√≠ch m√£) v√† tr·∫£ v·ªÅ ƒëi·ªÉm + g·ª£i √Ω s·ª≠a l·ªói.</i>"))
        grade_task.observe(on_task, names="value")
        on_task(None)

        def on_grade(_):
            with out_grade:
                clear_output()
                report = self.ai_grade_code(grade_task.value, code_in.value)
                display(report)

        btn_grade.on_click(on_grade)
        tab4 = widgets.VBox([grade_task, code_in, btn_grade, out_grade])

        tab3 = widgets.VBox([topic_dd, btn_q, out_q])

        subtabs = widgets.Tab(children=[tab1, tab2, tab3, tab4])
        subtabs.set_title(0, "H·ªèi ƒë√°p")
        subtabs.set_title(1, "T∆∞ v·∫•n thu·∫≠t to√°n")
        subtabs.set_title(2, "Quiz nhanh")
        subtabs.set_title(3, "Ch·∫•m b√†i")
        display(subtabs)



# MODULE 1: SORTING
    # MODULE 1: SORTING (ƒê√É N√ÇNG C·∫§P)
    def ui_sort(self, scr, ctl):
        with ctl:
            # 1. T·∫°o c√°c Widget nh·∫≠p li·ªáu
            lbl_input = widgets.Label("Nh·∫≠p s·ªë:")
            val_input = widgets.IntText(value=10, layout=widgets.Layout(width='70px'))

            # 2. T·∫°o c√°c n√∫t ch·ª©c nƒÉng thao t√°c d·ªØ li·ªáu
            btn_add = widgets.Button(description='Th√™m', button_style='success', icon='plus', layout=widgets.Layout(width='80px'))
            btn_del_last = widgets.Button(description='X√≥a cu·ªëi', button_style='warning', icon='arrow-left', layout=widgets.Layout(width='100px'))
            btn_del_val = widgets.Button(description='X√≥a s·ªë n√†y', button_style='danger', icon='trash', layout=widgets.Layout(width='100px'))
            btn_rand = widgets.Button(description='Random M·ªõi', button_style='info', icon='random')

            # 3. T·∫°o c√°c n√∫t thu·∫≠t to√°n (gi·ªØ nguy√™n)
            row1 = widgets.HBox([widgets.Button(description=x) for x in ['Bubble', 'Selection', 'Insertion']])
            row2 = widgets.HBox([widgets.Button(description=x) for x in ['Merge', 'Quick', 'Heap']])

            # --- X·ª¨ L√ù S·ª∞ KI·ªÜN ---

            # H√†m th√™m s·ªë v√†o m·∫£ng
            def f_add(b):
                self.sort_data.append(val_input.value)
                self.draw_sort(scr)

            # H√†m x√≥a s·ªë v·ª´a nh·∫≠p (x√≥a ph·∫ßn t·ª≠ cu·ªëi c√πng)
            def f_del_last(b):
                if self.sort_data:
                    self.sort_data.pop()
                    self.draw_sort(scr)

            # H√†m x√≥a gi√° tr·ªã b·∫•t k√¨ (x√≥a s·ªë ƒëang hi·ªÉn th·ªã trong √¥ nh·∫≠p)
            def f_del_val(b):
                val_to_remove = val_input.value
                if val_to_remove in self.sort_data:
                    self.sort_data.remove(val_to_remove) # Ch·ªâ x√≥a ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n t√¨m th·∫•y
                    self.draw_sort(scr)
                else:
                    # (T√πy ch·ªçn) C√≥ th·ªÉ in th√¥ng b√°o n·∫øu kh√¥ng t√¨m th·∫•y, nh∆∞ng ƒë·ªÉ ƒë∆°n gi·∫£n ta ch·ªâ v·∫Ω l·∫°i
                    pass

            # H√†m t·∫°o Random
            def f_rand(b):
                self.sort_data = random.sample(range(1,60), 7)
                self.draw_sort(scr)

            # G√°n s·ª± ki·ªán
            btn_add.on_click(f_add)
            btn_del_last.on_click(f_del_last)
            btn_del_val.on_click(f_del_val)
            btn_rand.on_click(f_rand)

            for b in row1.children + row2.children:
                b.on_click(lambda x, a=b.description: self.run_sort(scr, a))

            # --- HI·ªÇN TH·ªä GIAO DI·ªÜN ---
            # D√≤ng 1: Nh·∫≠p li·ªáu v√† Thao t√°c
            line_control = widgets.HBox([lbl_input, val_input, btn_add, btn_del_last, btn_del_val])

            display(widgets.VBox([
                line_control,       # D√≤ng nh·∫≠p li·ªáu
                btn_rand,           # N√∫t Random
                widgets.HTML("<hr>"), # ƒê∆∞·ªùng k·∫ª ngƒÉn c√°ch
                row1, row2          # C√°c n√∫t thu·∫≠t to√°n
            ]))

        self.draw_sort(scr)

    def draw_sort(self, scr, hl=[], done=[]):
        with scr:
            clear_output(wait=True); plt.figure(figsize=(7,3))
            c = ['#3498db']*len(self.sort_data)
            for i in hl: c[i]='#e74c3c'
            for i in done: c[i]='#2ecc71'
            plt.bar(range(len(self.sort_data)), self.sort_data, color=c)
            for i,v in enumerate(self.sort_data): plt.text(i,v+0.5,str(v),ha='center')
            plt.title("Sorting"); plt.show()

    def run_sort(self, scr, algo):
        d = self.sort_data
        n = len(d)

        # --- 1. BUBBLE SORT ---
        if algo == 'Bubble':
            for i in range(n):
                for j in range(0, n-i-1):
                    # Highlight 2 c·ªôt ƒëang so s√°nh (m√†u ƒë·ªè)
                    self.draw_sort(scr, hl=[j, j+1], done=list(range(n-i, n)))
                    time.sleep(0.3)

                    if d[j] > d[j+1]:
                        d[j], d[j+1] = d[j+1], d[j] # ƒê·ªïi ch·ªó
                        self.draw_sort(scr, hl=[j, j+1], done=list(range(n-i, n)))
                        time.sleep(0.3)

        # --- 2. SELECTION SORT ---
        elif algo == 'Selection':
            for i in range(n):
                min_idx = i
                # Highlight ph·∫ßn t·ª≠ ƒëang x√©t (m√†u ƒë·ªè)
                self.draw_sort(scr, hl=[i, min_idx], done=list(range(i)))
                time.sleep(0.3)

                for j in range(i+1, n):
                    # So s√°nh v·ªõi min hi·ªán t·∫°i
                    self.draw_sort(scr, hl=[i, j, min_idx], done=list(range(i)))
                    time.sleep(0.1)
                    if d[j] < d[min_idx]:
                        min_idx = j

                # ƒê·ªïi ch·ªó min v·ªÅ ƒë·∫ßu ƒëo·∫°n ch∆∞a x·∫øp
                d[i], d[min_idx] = d[min_idx], d[i]
                self.draw_sort(scr, hl=[i], done=list(range(i+1)))
                time.sleep(0.3)

        # --- 3. INSERTION SORT ---
        elif algo == 'Insertion':
            for i in range(1, n):
                key = d[i]
                j = i - 1
                # Di chuy·ªÉn c√°c ph·∫ßn t·ª≠ l·ªõn h∆°n key v·ªÅ sau
                while j >= 0 and d[j] > key:
                    self.draw_sort(scr, hl=[j, j+1], done=list(range(i))) # Highlight
                    time.sleep(0.2)
                    d[j+1] = d[j]
                    j -= 1
                    # C·∫≠p nh·∫≠t t·∫°m th·ªùi ƒë·ªÉ th·∫•y s·ª± d·ªãch chuy·ªÉn
                    d[j+1] = key
                    self.draw_sort(scr, hl=[j+1], done=list(range(i)))
                    time.sleep(0.2)
                d[j+1] = key

        # --- 4. QUICK SORT (ƒê·ªá quy) ---
        elif algo == 'Quick':
            def partition(low, high):
                pivot = d[high]
                i = low - 1
                for j in range(low, high):
                    self.draw_sort(scr, hl=[j, high], done=[])
                    time.sleep(0.1)
                    if d[j] < pivot:
                        i += 1
                        d[i], d[j] = d[j], d[i]
                d[i+1], d[high] = d[high], d[i+1]
                return i + 1

            def quick_sort_recursive(low, high):
                if low < high:
                    pi = partition(low, high)
                    quick_sort_recursive(low, pi - 1)
                    quick_sort_recursive(pi + 1, high)

            quick_sort_recursive(0, n-1)

        # --- C√ÅC SORT KH√ÅC (Merge, Heap) ---
        else:
            # Merge v√† Heap Sort kh√≥ bi·ªÉu di·ªÖn ƒë·∫πp tr√™n bi·ªÉu ƒë·ªì c·ªôt ƒë∆°n gi·∫£n
            # n√™n ta v·∫´n ƒë·ªÉ n√≥ ch·∫°y nhanh (ho·∫∑c b·∫°n c√≥ th·ªÉ b·ªï sung sau)
            with scr:
                clear_output(wait=True)
                print(f"‚ö° ƒêang ch·∫°y {algo} Sort (M√¥ ph·ªèng nhanh)...")
                time.sleep(1.0)
            self.sort_data.sort()

        # V·∫Ω l·∫°i k·∫øt qu·∫£ cu·ªëi c√πng (Full xanh l√°)
        self.draw_sort(scr, done=list(range(n)))

    # MODULE 2: LINKED LIST
    def ui_ll(self, scr, ctl):
        with ctl:
            val = widgets.IntText(value=99, layout=widgets.Layout(width='80px'))
            qval = widgets.IntText(value=20, layout=widgets.Layout(width='80px'))
            btns = [widgets.Button(description=d) for d in ['AddHead','AddTail','AddAfterQ','DelHead','Search','Trav']]
            lbl = widgets.Label("")

            def handler(b):
                d=b.description
                if d=='AddHead': self.ll_data.insert(0,val.value)
                elif d=='AddTail': self.ll_data.append(val.value)
                elif d=='DelHead':
                    if self.ll_data: self.ll_data.pop(0)
                elif d=='AddAfterQ':
                    if qval.value in self.ll_data: self.ll_data.insert(self.ll_data.index(qval.value)+1, val.value)
                    else: lbl.value = "Kh√¥ng t√¨m th·∫•y Q"
                elif d=='Search':
                    for i,v in enumerate(self.ll_data):
                        self.draw_ll(scr, hl=i); time.sleep(0.3)
                        if v==val.value: lbl.value="T√¨m th·∫•y!"; return
                    lbl.value="Kh√¥ng th·∫•y"
                elif d=='Trav':
                    for i in range(len(self.ll_data)): self.draw_ll(scr, hl=i); time.sleep(0.3)
                self.draw_ll(scr)

            for b in btns: b.on_click(handler)
            display(widgets.HBox([val, qval, lbl]), widgets.HBox(btns[:3]), widgets.HBox(btns[3:]))
        self.draw_ll(scr)

    def draw_ll(self, scr, hl=-1):
        with scr:
            clear_output(wait=True); G=nx.DiGraph(); pos={}
            for i,v in enumerate(self.ll_data): G.add_node(i, label=str(v)); pos[i]=(i,0);
            for i in range(len(self.ll_data)-1): G.add_edge(i,i+1)
            c=['orange' if i==hl else 'white' for i in range(len(self.ll_data))]
            plt.figure(figsize=(8,2)); nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G,'label'), node_size=1000, node_color=c, edgecolors='black'); plt.show()

    # MODULE 3: BST
    def bst_insert(self, root, val):
        if not root: return TreeNode(val)
        if val < root.val: root.left = self.bst_insert(root.left, val)
        else: root.right = self.bst_insert(root.right, val)
        return root

    # MODULE 3: BST (ƒê√É S·ª¨A BFS/DFS)
    def ui_bst(self, scr, ctl):
        with ctl:
            # 1. T·∫°o widget nh·∫≠p li·ªáu
            val = widgets.IntText(value=45, layout=widgets.Layout(width='80px'))

            # 2. T·∫°o c√°c n√∫t ch·ª©c nƒÉng
            btns_edit = [
                widgets.Button(description='Insert', button_style='success'),
                widgets.Button(description='Delete', button_style='danger'),
                widgets.Button(description='Random', button_style='info')
            ]
            btns_algo = [
                widgets.Button(description='Search'),
                widgets.Button(description='BFS'), # Duy·ªát chi·ªÅu r·ªông
                widgets.Button(description='DFS')  # Duy·ªát chi·ªÅu s√¢u
            ]

            # 3. X·ª≠ l√Ω s·ª± ki·ªán
            def handler(b):
                d = b.description

                # --- Nh√≥m ch·ªânh s·ª≠a c√¢y ---
                if d == 'Insert':
                    self.bst_root = self.bst_insert(self.bst_root, val.value)

                elif d == 'Delete':
                    # ƒê·ªÉ ƒë∆°n gi·∫£n ho√° demo, ta t·∫°o l·∫°i c√¢y m·ªõi tr·ª´ node c·∫ßn x√≥a
                    # (X√≥a ƒë√∫ng chu·∫©n BST kh√° d√†i d√≤ng)
                    if self.bst_root:
                        nodes = []
                        def get_nodes(n):
                            if n: nodes.append(n.val); get_nodes(n.left); get_nodes(n.right)
                        get_nodes(self.bst_root)
                        if val.value in nodes: nodes.remove(val.value)

                        self.bst_root = None
                        for x in nodes: self.bst_root = self.bst_insert(self.bst_root, x)

                elif d == 'Random':
                    self.bst_root = None
                    # T·∫°o ng·∫´u nhi√™n 7 node
                    for x in random.sample(range(1, 100), 7):
                        self.bst_root = self.bst_insert(self.bst_root, x)

                # --- Nh√≥m thu·∫≠t to√°n duy·ªát ---
                elif d == 'BFS':
                    if not self.bst_root:
                        with scr: print("‚ö†Ô∏è C√¢y r·ªóng!"); time.sleep(1)
                    else:
                        queue = [self.bst_root]
                        visited = [] # Danh s√°ch c√°c node ƒë√£ duy·ªát

                        while queue:
                            curr = queue.pop(0) # L·∫•y ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n (FIFO)
                            visited.append(curr.val)

                            # V·∫Ω highlight T·∫§T C·∫¢ c√°c node ƒë√£ ƒëi qua ƒë·ªÉ th·∫•y ƒë∆∞·ªùng ƒëi
                            self.draw_bst(scr, hl=visited)
                            time.sleep(0.5) # Ch·∫≠m l·∫°i ƒë·ªÉ nh√¨n k·ªãp

                            if curr.left: queue.append(curr.left)
                            if curr.right: queue.append(curr.right)

                elif d == 'DFS':
                    visited = []
                    def dfs_rec(node):
                        if not node: return

                        # X·ª≠ l√Ω node (Pre-order)
                        visited.append(node.val)
                        self.draw_bst(scr, hl=visited)
                        time.sleep(0.5)

                        dfs_rec(node.left)
                        dfs_rec(node.right)

                    if not self.bst_root:
                        with scr: print("‚ö†Ô∏è C√¢y r·ªóng!"); time.sleep(1)
                    else:
                        dfs_rec(self.bst_root)

                elif d == 'Search':
                    curr = self.bst_root
                    path = []
                    found = False
                    while curr:
                        path.append(curr.val)
                        self.draw_bst(scr, hl=path)
                        time.sleep(0.5)

                        if curr.val == val.value:
                            found = True
                            with scr: print(f"‚úÖ ƒê√£ t√¨m th·∫•y {val.value}!"); time.sleep(1)
                            break

                        if val.value < curr.val: curr = curr.left
                        else: curr = curr.right

                    if not found:
                        with scr: print(f"‚ùå Kh√¥ng t√¨m th·∫•y {val.value}"); time.sleep(1)

                # V·∫Ω l·∫°i tr·∫°ng th√°i cu·ªëi c√πng
                self.draw_bst(scr)

            # G√°n s·ª± ki·ªán click
            for b in btns_edit + btns_algo:
                b.on_click(handler)

            # Hi·ªÉn th·ªã ra m√†n h√¨nh
            display(widgets.HBox([widgets.Label("Node:"), val]),
                    widgets.HBox(btns_edit),
                    widgets.HBox(btns_algo))

        # V·∫Ω l·∫ßn ƒë·∫ßu khi m·ªõi m·ªü tab
        self.draw_bst(scr)

    def draw_bst(self, scr, hl=[]):
        with scr:
            clear_output(wait=True); G=nx.DiGraph(); pos={}
            def build(n, x=0, y=0, w=4):
                if n:
                    pos[n.val]=(x,y); G.add_node(n.val)
                    if n.left: G.add_edge(n.val, n.left.val); build(n.left, x-w/2, y-1, w/2)
                    if n.right: G.add_edge(n.val, n.right.val); build(n.right, x+w/2, y-1, w/2)
            if self.bst_root: build(self.bst_root)
            c=['orange' if n in hl else 'white' for n in G.nodes()]
            if G.nodes: plt.figure(figsize=(6,4)); nx.draw(G, pos, with_labels=True, node_color=c, edgecolors='black'); plt.show()

    # MODULE 4: GRAPH
    # ==========================================
    # MODULE 4: GRAPH (N√ÇNG C·∫§P - CHI TI·∫æT)
    # ==========================================
    def ui_graph(self, scr, ctl):
        # C·ªë ƒë·ªãnh layout ƒë·ªÉ node kh√¥ng nh·∫£y khi v·∫Ω l·∫°i
        if not hasattr(self, 'graph_pos'):
            self.graph_pos = nx.spring_layout(self.G, seed=42)

        with ctl:
            lbl_algo = widgets.HTML("<b>Ch·ªçn thu·∫≠t to√°n:</b>")

            nodes = list(self.G.nodes())
            if not nodes:
                nodes = ['A']
            dd_start = widgets.Dropdown(options=nodes, value=nodes[0], description='Start:')

            btn_bfs = widgets.Button(description='BFS', button_style='success')
            btn_dfs = widgets.Button(description='DFS', button_style='success')
            btn_dijkstra = widgets.Button(description='Dijkstra (ƒê∆∞·ªùng ƒëi)', button_style='primary')
            btn_prim = widgets.Button(description='Prim (MST)', button_style='info')
            btn_reset = widgets.Button(description='Reset Graph', button_style='warning')

            # Output areas
            self.log_output = widgets.Output(layout={'border':'1px solid #ddd', 'height':'200px', 'overflow_y':'scroll', 'padding':'5px'})
            self.table_output = widgets.Output(layout={'margin-top':'10px'})

            # --- Event handlers ---
            def run_bfs_vis(b):
                self.run_graph_traversal(scr, dd_start.value, mode='bfs')

            def run_dfs_vis(b):
                self.run_graph_traversal(scr, dd_start.value, mode='dfs')

            def run_dijkstra_vis(b):
                self.visualize_dijkstra(scr)

            def run_prim_vis(b):
                self.visualize_prim(scr)

            def reset_graph(b):
                with self.log_output:
                    clear_output()
                    print("ƒê√£ reset ƒë·ªì th·ªã.")
                with self.table_output:
                    clear_output()
                self.draw_graph_step(scr)

            btn_bfs.on_click(run_bfs_vis)
            btn_dfs.on_click(run_dfs_vis)
            btn_dijkstra.on_click(run_dijkstra_vis)
            btn_prim.on_click(run_prim_vis)
            btn_reset.on_click(reset_graph)

            display(
                widgets.VBox([
                    widgets.HBox([lbl_algo, dd_start]),
                    widgets.HBox([btn_bfs, btn_dfs, btn_dijkstra, btn_prim, btn_reset]),
                    widgets.HTML("<hr><b>üìú Nh·∫≠t k√Ω thu·∫≠t to√°n:</b>"),
                    self.log_output,
                    widgets.HTML("<b>üìä B·∫£ng kho·∫£ng c√°ch D[u]:</b>"),
                    self.table_output
                ])
            )

        # V·∫Ω tr·∫°ng th√°i ban ƒë·∫ßu
        self.draw_graph_step(scr)
    def run_graph_traversal(self, scr, start, mode='bfs'):
        """Minh ho·∫° BFS/DFS tr√™n ƒë·ªì th·ªã hi·ªán t·∫°i (offline, t·ª´ng b∆∞·ªõc)."""
        if start not in self.G:
            return
        visited = []
        if mode == 'bfs':
            from collections import deque
            q = deque([start])
            seen = set([start])
            while q:
                u = q.popleft()
                visited.append(u)
                for v in self.G.neighbors(u):
                    if v not in seen:
                        seen.add(v); q.append(v)
        else:  # dfs
            stack = [start]
            seen = set()
            while stack:
                u = stack.pop()
                if u in seen:
                    continue
                seen.add(u)
                visited.append(u)
                # ƒë·∫£o ng∆∞·ª£c ƒë·ªÉ nh√¨n "gi·ªëng s√°ch gi√°o khoa" h∆°n
                neigh = sorted(list(self.G.neighbors(u)), reverse=True)
                for v in neigh:
                    if v not in seen:
                        stack.append(v)

        # animate
        for i, u in enumerate(visited, 1):
            cmap = []
            for n in self.G.nodes():
                if n == u:
                    cmap.append('#f1c40f')  # current
                elif n in visited[:i]:
                    cmap.append('#2ecc71')  # visited
                else:
                    cmap.append('#ecf0f1')  # default
            self.draw_graph_step(scr, color_map=cmap, title=f"{mode.upper()} step {i}/{len(visited)}: visit {u}")
            plt.pause(0.6)

        with self.log_output:
            print(f"{mode.upper()} order:", ' -> '.join(visited))

    def draw_graph_step(self, scr, color_map=None, edge_colors=None, title="Graph"):
        with scr:
            clear_output(wait=True)
            plt.figure(figsize=(6, 4))

            # M·∫∑c ƒë·ªãnh m√†u tr·∫Øng n·∫øu kh√¥ng c√≥ map
            if color_map is None:
                color_map = ['#ecf0f1'] * len(self.G.nodes())

            # V·∫Ω Node
            nx.draw_networkx_nodes(self.G, self.graph_pos, node_color=color_map, node_size=700, edgecolors='black')
            nx.draw_networkx_labels(self.G, self.graph_pos)

            # V·∫Ω Edge (C·∫°nh)
            if edge_colors is None:
                edge_colors = ['black'] * len(self.G.edges())

            nx.draw_networkx_edges(self.G, self.graph_pos, edge_color=edge_colors, width=2)

            # V·∫Ω Tr·ªçng s·ªë (Weight)
            labels = nx.get_edge_attributes(self.G, 'weight')
            nx.draw_networkx_edge_labels(self.G, self.graph_pos, edge_labels=labels)

            plt.title(title)
            plt.axis('off')
            plt.show()

    # --- H√ÄM LOGGING ---
    def log(self, msg):
        with self.log_output:
            print(msg)

    def show_dist_table(self, dist):
        with self.table_output:
            clear_output(wait=True)
            # T·∫°o HTML table ƒë∆°n gi·∫£n
            html = "<table border='1' style='width:100%; border-collapse:collapse; text-align:center;'>"
            html += "<tr style='background:#f4f4f4'><th>ƒê·ªânh</th>" + "".join([f"<th>{n}</th>" for n in dist]) + "</tr>"
            html += "<tr><td><b>Dist</b></td>" + "".join([f"<td>{dist[n] if dist[n]!=999 else '‚àû'}</td>" for n in dist]) + "</tr>"
            html += "</table>"
            display(HTML(html))

    # --- THU·∫¨T TO√ÅN 1: DIJKSTRA VISUALIZATION ---
    def visualize_dijkstra(self, scr):
        start_node = 'A'
        nodes = list(self.G.nodes())
        dist = {n: 999 for n in nodes} # 999 ƒë·∫°i di·ªán v√¥ c·ª±c
        dist[start_node] = 0
        parent = {n: None for n in nodes}
        visited = set()

        with self.log_output: clear_output(); print(f"üèÅ B·∫Øt ƒë·∫ßu Dijkstra t·ª´ ƒë·ªânh {start_node}...")
        self.show_dist_table(dist)

        # Loop ch√≠nh (M√¥ ph·ªèng Priority Queue b·∫±ng c√°ch ch·ªçn min th·ªß c√¥ng ƒë·ªÉ d·ªÖ visual)
        for _ in range(len(nodes)):
            # 1. T√¨m ƒë·ªânh ch∆∞a thƒÉm c√≥ dist nh·ªè nh·∫•t
            u = None
            min_val = 999
            for n in nodes:
                if n not in visited and dist[n] < min_val:
                    min_val = dist[n]
                    u = n

            if u is None or dist[u] == 999: break # Kh√¥ng c√≤n ƒë∆∞·ªùng ƒëi

            # VISUAL: T√¥ m√†u ƒê·ªè cho ƒë·ªânh ƒëang x√©t
            visited.add(u)
            colors = ['#e74c3c' if n == u else ('#2ecc71' if n in visited else '#ecf0f1') for n in nodes]
            self.log(f"üëâ ƒêang x√©t ƒë·ªânh {u} (Dist={dist[u]})")
            self.draw_graph_step(scr, color_map=colors, title=f"ƒêang x√©t: {u}")
            time.sleep(1.0)

            # 2. Duy·ªát c√°c h√†ng x√≥m
            for v, data in self.G[u].items():
                if v in visited: continue

                weight = data['weight']
                new_dist = dist[u] + weight

                # VISUAL: T√¥ m√†u V√†ng cho h√†ng x√≥m ƒëang ki·ªÉm tra
                colors_check = list(colors)
                colors_check[nodes.index(v)] = '#f1c40f' # V√†ng
                self.draw_graph_step(scr, color_map=colors_check, title=f"Ki·ªÉm tra c·∫°nh {u}->{v} (w={weight})")
                time.sleep(0.5)

                if new_dist < dist[v]:
                    self.log(f"   ‚ö° C·∫≠p nh·∫≠t {v}: {dist[v]} ‚ûù {new_dist}")
                    dist[v] = new_dist
                    parent[v] = u
                    self.show_dist_table(dist)
                    time.sleep(0.5)
                else:
                    self.log(f"   StartGi·ªØ nguy√™n {v} (V√¨ {new_dist} >= {dist[v]})")

        # K·∫æT TH√öC: V·∫Ω ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t ƒë·∫øn ƒë√≠ch (v√≠ d·ª• D)
        path = []
        curr = 'D'
        if dist[curr] != 999:
            while curr:
                path.append(curr)
                curr = parent[curr]
            path.reverse() # ƒê·∫£o ng∆∞·ª£c ƒë·ªÉ c√≥ A -> ... -> D

            self.log(f"‚úÖ ƒê∆∞·ªùng ƒëi ng·∫Øn nh·∫•t A->D: {' -> '.join(path)} (Chi ph√≠: {dist['D']})")

            # T√¥ m√†u ƒë∆∞·ªùng ƒëi cu·ªëi c√πng (M√†u xanh d∆∞∆°ng)
            final_colors = ['#3498db' if n in path else '#ecf0f1' for n in nodes]

            # T√¥ m√†u c·∫°nh thu·ªôc ƒë∆∞·ªùng ƒëi
            edge_colors = []
            path_edges = list(zip(path, path[1:]))
            for u, v in self.G.edges():
                if (u, v) in path_edges or (v, u) in path_edges:
                    edge_colors.append('red')
                else:
                    edge_colors.append('#ddd')

            self.draw_graph_step(scr, color_map=final_colors, edge_colors=edge_colors, title="HO√ÄN T·∫§T")

    # --- THU·∫¨T TO√ÅN 2: PRIM VISUALIZATION (MST) ---
    def visualize_prim(self, scr):
        start_node = 'A'
        mst_edges = []
        visited = {start_node}
        edges = list(self.G.edges(data=True))

        nodes = list(self.G.nodes())

        with self.log_output: clear_output(); print(f"üå≤ B·∫Øt ƒë·∫ßu Prim t·ª´ ƒë·ªânh {start_node}...")

        while len(visited) < len(nodes):
            min_edge = None
            min_weight = 999

            # T√¨m c·∫°nh nh·ªè nh·∫•t n·ªëi t·ª´ t·∫≠p ƒë√£ thƒÉm ra ngo√†i
            for u, v, data in edges:
                w = data['weight']
                # ƒêi·ªÅu ki·ªán: 1 ƒë·ªânh ƒë√£ thƒÉm, 1 ƒë·ªânh ch∆∞a thƒÉm
                if (u in visited and v not in visited) or (v in visited and u not in visited):

                    # VISUAL: Highlight c·∫°nh ƒëang c√¢n nh·∫Øc (V√†ng nh·∫°t)
                    # (Ph·∫ßn n√†y n·∫øu v·∫Ω li√™n t·ª•c s·∫Ω r·∫•t ch·∫≠m, n√™n ta ch·ªâ v·∫Ω khi t√¨m th·∫•y min m·ªõi ho·∫∑c ch·ªët)
                    if w < min_weight:
                        min_weight = w
                        min_edge = (u, v, w)

            if min_edge:
                u, v, w = min_edge
                new_node = v if u in visited else u
                visited.add(new_node)
                mst_edges.append((u, v))

                self.log(f"üîó Ch·ªçn c·∫°nh {u}-{v} (weight={w}) v√†o MST.")

                # VISUAL: V·∫Ω tr·∫°ng th√°i hi·ªán t·∫°i
                # Node ƒë√£ thƒÉm: Xanh l√°, Node m·ªõi: Cam
                node_colors = ['#2ecc71' if n in visited else '#ecf0f1' for n in nodes]

                # C·∫°nh trong MST: ƒê·ªè ƒë·∫≠m
                edge_colors = []
                for eu, ev in self.G.edges():
                    if (eu, ev) in mst_edges or (ev, eu) in mst_edges:
                        edge_colors.append('#e74c3c')
                    else:
                        edge_colors.append('#ddd')

                self.draw_graph_step(scr, color_map=node_colors, edge_colors=edge_colors, title=f"ƒê√£ th√™m {u}-{v}")
                time.sleep(1.0)
            else:
                break

        self.log("‚úÖ Ho√†n th√†nh C√¢y khung nh·ªè nh·∫•t (MST).")

    # OTHER MODULES
    def build_theory(self):
        with self.out_theory:
            acc = widgets.Accordion(children=[widgets.HTML(THEORY_CONTENT[k]) for k in THEORY_CONTENT])
            for i,t in enumerate(THEORY_CONTENT): acc.set_title(i,t)
            display(acc)
    def build_quiz(self):
        with self.out_quiz:
            display(HTML("<h3>üìù B√ÄI KI·ªÇM TRA TR·∫ÆC NGHI·ªÜM (10 C√ÇU)</h3>"))
            self.quiz_widgets = []

            for i, q in enumerate(QUIZ_DB):
                lbl = widgets.HTML(f"<b>C√¢u {i+1}:</b> {q['q']}")
                rdo = widgets.RadioButtons(options=q['opts'], index=None)
                self.quiz_widgets.append((rdo, q['a']))
                display(lbl, rdo, HTML("<hr>"))

            btn_sub = widgets.Button(description='N·ªôp B√†i & Ch·∫•m ƒêi·ªÉm', button_style='success', layout=widgets.Layout(width='300px'))
            out_res = widgets.Output()

            def check(b):
                score = 0
                with out_res:
                    clear_output()
                    print("--- K·∫æT QU·∫¢ CHI TI·∫æT ---")
                    for i, (rdo, ans) in enumerate(self.quiz_widgets):
                        if rdo.value == ans:
                            score += 1
                            print(f"C√¢u {i+1}: ‚úÖ ƒê√∫ng")
                        else:
                            print(f"C√¢u {i+1}: ‚ùå Sai (ƒê√°p √°n ƒë√∫ng: {ans})")
                    print("-" * 30)
                    display(HTML(f"<h3 style='color: {'green' if score >= 8 else 'red'}'>T·ªîNG ƒêI·ªÇM: {score}/10</h3>"))

            btn_sub.on_click(check)
            display(btn_sub, out_res)

    # Assignments
    def build_assign(self):
        with self.out_assign:
            display(HTML("<h3>üíª LAB: B√ÄI T·∫¨P TH·ª∞C H√ÄNH</h3>"))

            # --- B√ÄI 1: SORTING (ƒê√É S·ª¨A) ---
            # 1. T·∫°o c√°c widget v√† g√°n v√†o self ƒë·ªÉ h√†m kh√°c g·ªçi ƒë∆∞·ª£c
            lbl_req = widgets.HTML("<b>Y√™u c·∫ßu:</b> C√†i ƒë·∫∑t thu·∫≠t to√°n Bubble Sort b·∫±ng Python.")

            self.code_input_sort = widgets.Textarea(
                placeholder='def bubble_sort(arr):...',
                layout=widgets.Layout(width='100%', height='150px')
            )

            self.btn_submit_sort = widgets.Button(
                description='N·ªôp b√†i Sorting',
                button_style='primary'
            )

            # Quan tr·ªçng: Widget n√†y ƒë·ªÉ hi·ªán k·∫øt qu·∫£ check b√†i
            self.result_sort_output = widgets.Output(layout={'border': '1px solid #ddd', 'padding': '10px', 'margin-top': '5px'})

            # 2. Gom v√†o VBox
            t1 = widgets.VBox([
                lbl_req,
                self.code_input_sort,
                self.btn_submit_sort,
                self.result_sort_output
            ])

            # 3. H√†m x·ª≠ l√Ω logic ch·∫•m b√†i
            def check_sort_solution(b):
                with self.result_sort_output:
                    self.result_sort_output.clear_output()
                    print("‚è≥ ƒêang ch·∫•m b√†i...")

                    try:
                        user_code = self.code_input_sort.value
                        if not user_code.strip():
                            print("‚ö†Ô∏è B·∫°n ch∆∞a nh·∫≠p code n√†o c·∫£!")
                            return

                        # T·∫°o m√¥i tr∆∞·ªùng ch·∫°y code
                        local_scope = {}
                        exec(user_code, {}, local_scope)

                        if 'bubble_sort' not in local_scope:
                            print("‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y h√†m 'bubble_sort'.")
                            print("   üëâ H√£y ƒë·ªãnh nghƒ©a: def bubble_sort(arr):")
                            return

                        test_arr = [64, 34, 25, 12, 22, 11, 90]
                        expected = sorted(test_arr)

                        # Copy m·∫£ng ƒë·ªÉ test
                        user_func = local_scope['bubble_sort']
                        arr_copy = list(test_arr)
                        user_func(arr_copy)

                        if arr_copy == expected:
                            print(f"‚úÖ CH√çNH X√ÅC! K·∫øt qu·∫£: {arr_copy}")
                        else:
                            print(f"‚ùå SAI. K·∫øt qu·∫£ c·ªßa b·∫°n: {arr_copy}")
                            print(f"   üëâ K·∫øt qu·∫£ ƒë√∫ng: {expected}")

                    except Exception as e:
                        print(f"üí• L·ªói code: {e}")

            # 4. G√°n s·ª± ki·ªán click (L√∫c n√†y self.btn_submit_sort ƒë√£ t·ªìn t·∫°i)
            self.btn_submit_sort.on_click(check_sort_solution)

            # --- B√ÄI 2 & 3 (GI·ªÆ NGUY√äN HO·∫∂C S·ª¨A T∆Ø∆†NG T·ª∞ N·∫æU C·∫¶N) ---
            t2 = widgets.VBox([
                widgets.HTML("<b>Y√™u c·∫ßu:</b> C√†i ƒë·∫∑t h√†m <code>add_head(node)</code>."),
                widgets.Textarea(placeholder='def add_head(self, data):...', layout=widgets.Layout(width='100%', height='150px')),
                widgets.Button(description='N·ªôp b√†i Linked List', button_style='primary')
            ])

            t3 = widgets.VBox([
                widgets.HTML("<b>Y√™u c·∫ßu:</b> C√†i ƒë·∫∑t h√†m duy·ªát c√¢y BFS."),
                widgets.Textarea(placeholder='def bfs(root):...', layout=widgets.Layout(width='100%', height='150px')),
                widgets.Button(description='N·ªôp b√†i BST', button_style='primary')
            ])

            # T·∫°o Accordion ch·ª©a c√°c b√†i t·∫≠p
            acc = widgets.Accordion(children=[t1, t2, t3])
            acc.set_title(0, 'Lab 1: Sorting Algorithms')
            acc.set_title(1, 'Lab 2: Linked List Implementation')
            acc.set_title(2, 'Lab 3: Tree Traversal')
            display(acc)



    # ƒê·ªÅ thi
    def build_exam(self):
        with self.out_exam:
            display(HTML("<h3>üî• ƒê·ªÄ THI CU·ªêI K·ª≤ (NG·∫™U NHI√äN)</h3>"))
            display(HTML("<i>H·ªá th·ªëng s·∫Ω ch·ªçn ng·∫´u nhi√™n 5 c√¢u h·ªèi t·ª´ ng√¢n h√†ng ƒë·ªÅ thi ƒë·ªÉ t·∫°o ƒë·ªÅ cho b·∫°n.</i>"))

            btn_gen = widgets.Button(description='üé≤ T·∫†O ƒê·ªÄ THI M·ªöI', button_style='danger', layout=widgets.Layout(width='200px', height='50px'))
            out_paper = widgets.Output(layout={'border': '3px solid #333', 'padding': '20px', 'margin-top': '20px', 'background': '#fff'})

            def gen_exam(b):
                with out_paper:
                    clear_output()
                    questions = random.sample(EXAM_QUESTION_POOL, 5)
                    display(HTML(f"<h3 style='text-align:center'>ƒê·ªÄ THI M√É S·ªê: {random.randint(1000,9999)}</h3><hr>"))
                    for i, q in enumerate(questions):
                        display(HTML(f"<p><b>C√¢u {i+1}:</b> {q}</p>"))
                        display(widgets.Textarea(placeholder='L√†m b√†i t·∫°i ƒë√¢y...', layout=widgets.Layout(width='100%', height='80px')))
                    display(HTML("<hr><center><i>-- H·∫øt --</i></center>"))
            btn_gen.on_click(gen_exam)
            display(btn_gen, out_paper)

def main():
    VisualGo()


if __name__ == "__main__":
    main()
```
